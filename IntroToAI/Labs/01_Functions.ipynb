## 1. Functions as Black Boxes

A **function** is a named sequence of instructions. You give it inputs (called **arguments**), it performs a task, and it may send back a result (called a **return value**).

```
arguments  ──►  [ function ]  ──►  return value
```

The key idea is that you **do not need to know how** a function works internally — only what it expects and what it produces.

```python
price = round(6.8275, 2)   # arguments: 6.8275 and 2
print(price)               # 6.83  ← return value
```

## 2. Defining Functions with `def`

Use the `def` keyword to create your own function. The body must be indented.

````python
def cube_volume(side_length):
    volume = side_length ** 3
    return volume

print(cube_volume(2))   # 8
print(cube_volume(10))  # 1000
````

**Vocabulary recap:**

| Term | Meaning | Example |
|---|---|---|
| Function name | Label you use to call the function | `cube_volume` |
| Parameter | Variable in the function header | `side_length` |
| Argument | Actual value passed when calling | `2` |
| `return` | Sends a result back to the caller | `return volume` |

### 2.1 Program Flow

Python **stores** the function definition when it reads `def ...`, but **does not run the body** until the function is called.

Execution order for `result = cube_volume(4)`:
1. Python reaches the call `cube_volume(4)`
2. Parameter `side_length` is created and set to `4`
3. Body runs: `volume = 4 ** 3` → `64`
4. `return volume` sends `64` back
5. `result` holds `64`; execution continues

````python
def area(width, height):
    return width * height

room = area(4, 6)
print(room)  # 24
````

### 💡 Example 1
A function `square_area(side)` that returns the area of a square, tested with two different values.

````python
def square_area(side):
    return side ** 2

print(square_area(5))   # 25
print(square_area(12))  # 144
````

## 3. Positional vs. Keyword Arguments

Arguments can be matched by **position** (order matters) or by **name** (keyword arguments).

````python
def describe_student(name, major, year):
    return f"{name} studies {major} in year {year}."

# Positional — order must match the definition
print(describe_student("Aisha", "AI", 2))

# Keyword — order does not matter
print(describe_student(name="Aisha", major="AI", year=2))
print(describe_student(year=2, name="Aisha", major="AI"))
````

### Named arguments — a readability win

Compare these two calls:

```python
compute_profit(30000, 1500, 7500, 910)                               # hard to read
compute_profit(sales=30000, refunds=1500, costs=7500, expenses=910)  # clear!
```

> **Tip:** Use keyword arguments whenever you pass several numbers or similarly-typed values.

````python
def say_name(name, times):
    for _ in range(times):
        print(name)

say_name('John', 3)             # positional
say_name('John', times=3)       # one keyword arg
say_name(name='John', times=3)  # both keyword
````

## 4. Default Parameter Values

A **default value** is used when the caller does not supply that argument.

````python
def greet(name, language="English"):
    if language == "Arabic":
        return f"Ahlan, {name}!"
    return f"Hello, {name}!"

print(greet("Sara"))                    # uses default
print(greet("Sara", language="Arabic")) # override
````

````python
def say_name(name, times=3):
    for _ in range(times):
        print(name)

say_name('Mary')    # times defaults to 3
say_name('Mary', 1) # override to 1
````

> **Rule:** Parameters with defaults must come **after** parameters without defaults.

### 💡 Example 2
A function `power(base, exponent=2)` that returns `base` raised to `exponent`, using a default argument.
- `power(3)` → `9`
- `power(3, 3)` → `27`

````python
def power(base, exponent=2):
    return base ** exponent

print(power(3))     # 9  — exponent defaults to 2
print(power(3, 3))  # 27 — exponent overridden to 3
````

## 5. `return` vs. `print` — and Side Effects

````python
# Returns a value — result can be stored and reused
def add(a, b):
    return a + b

x = add(2, 3)     # x = 5
y = add(x, 10)    # y = 15
print(y)
````

````python
# Side effect — just prints; cannot store the result
def print_total(a, b):
    print(a + b)

result = print_total(2, 3)  # prints 5...
print(result)               # ...but result is None!
````

**Pure function:** depends only on its inputs, returns a result, changes nothing else. Easier to test.

**Side effect:** the function changes something outside itself (prints, writes a file, modifies a list). Sometimes necessary — but should be intentional.

## 7. Variable Scope — The LEGB Rule

When Python sees a name it searches in this order:

```
L → Local      (inside the current function)
E → Enclosing  (outer function, if nested)
G → Global     (module level)
B → Built-in   (len, print, range, …)
```

### 7.1 Local variables stay inside their function

````python
def scope_test():
    x = 1  # local to scope_test

scope_test()
# print(x)  # ← NameError: x is not defined out here
````

### 7.2 A function can *read* a global variable

````python
some_global_var = 'Hello World!'

def scope_test():
    print("Inside fun: ", some_global_var)  # readable ✅

scope_test()
print("Outside fun: ", some_global_var)
````

### 7.3 Assigning inside a function creates a *new* local variable

````python
some_global_var = 'Hello World!'

def scope_test():
    some_global_var = 'Goodbye cruel World!'  # new LOCAL variable
    print("Inside fun: ", some_global_var)

scope_test()
print("Outside code: ", some_global_var)  # original unchanged
````

### 7.4 Cannot modify a global variable without declaring it

````python
age = 13

def scope_test():
    # age += 1  # ← UnboundLocalError!
    # Python sees the assignment and treats 'age' as local,
    # then tries to read it before it has been assigned.
    pass

scope_test()
````

### 7.5 `global` — explicitly modifying a global variable

````python
age = 13

def scope_test():
    global age          # 😡 works, but NOT recommended
    age += 1
    print("Inside fun: ", age)

scope_test()
print("Outside fun: ", age)
````

> **Warning:** Frequent use of `global` makes code harder to follow. Prefer passing values in and returning results out.

## 8. Nested Functions & `nonlocal`

A function defined **inside another function** is called a **nested (inner) function**. It is only visible within its enclosing function — this is the **E** in LEGB.

````python
def calculate(op, x, y):
    def add():      return x + y
    def subtract(): return x - y
    def multiply(): return x * y
    def divide():   return x / y

    if   op == 'add':      return add()
    elif op == 'subtract': return subtract()
    elif op == 'multiply': return multiply()
    else:                  return divide()

print(calculate('add', 2, 3))       # 5
print(calculate('multiply', 4, 5))  # 20
print(calculate('divide', 10, 4))   # 2.5
````

The inner functions access `x` and `y` from the **enclosing** scope without needing them as parameters.

### 8.1 `nonlocal` — modifying an enclosing variable

Just as `global` targets module-level variables, `nonlocal` targets a variable from the **enclosing function**.

````python
result = 'Something completely unrelated'

def calculate(op, x, y):
    result = 'not changed yet'    # enclosing-scope variable

    def add():
        nonlocal result           # modifies enclosing 'result', NOT the global one
        result = x + y

    def subtract():
        nonlocal result
        result = x - y

    if op == 'add': add()
    else:           subtract()

    return result

print("Calc result: ",  calculate('add', 2, 3))  # 5
print("Global result: ", result)                  # still 'Something completely unrelated'
````

**Scope keyword summary:**

| Keyword | Targets | Recommendation |
|---|---|---|
| *(none)* | Creates a new local variable | Default choice |
| `global` | Module-level variable | Avoid when possible |
| `nonlocal` | Enclosing function variable | Useful in specific nested patterns |

### 💡 Example 3
`make_counter` returns an inner function that increases a private counter each time it is called. `nonlocal` lets the inner function modify the enclosing `count` variable without touching any global.

````python
def make_counter():
    count = 0
    def increment():
        nonlocal count
        count += 1
        return count
    return increment

counter = make_counter()
print(counter())  # 1
print(counter())  # 2
print(counter())  # 3
````

## 9. Putting It All Together — Worked Example

````python
def compute_profit(sales, refunds=0, costs=0, expenses=0):
    """Calculate net profit.

    Parameters
    ----------
    sales    : float — total revenue
    refunds  : float — amount refunded (default 0)
    costs    : float — cost of goods (default 0)
    expenses : float — operating expenses (default 0)

    Returns
    -------
    float — net profit
    """
    return sales - refunds - costs - expenses


# Hard-to-read positional call
print(compute_profit(30000, 1500, 7500, 910))

# Clear keyword call
print(compute_profit(sales=30000, refunds=1500, costs=7500, expenses=910))
````

## 10. Quick-Reference Cheat Sheet

```python
# Defining
def greet(name, times=1):          # default argument
    """Print name 'times' times."""
    for _ in range(times):
        print(f"Hello, {name}!")

# Calling
greet("Lena")                      # positional, times uses default
greet("Lena", times=3)             # keyword argument
greet(name="Lena", times=3)        # all keyword

# Scope
x = 10                             # global
def f():
    x = 99                         # local (shadows global)

def g():
    global x
    x = 99                         # modifies global (avoid!)

# Nested + nonlocal
def outer():
    count = 0
    def inner():
        nonlocal count
        count += 1
    inner()
    return count
```

## 11. More Examples

````python
# Example 4 — side effect function (no return value)
# box_string(contents) prints a string surrounded by a border.

def box_string(contents):
    n = len(contents)
    border = '-' * (n + 2)
    print(border)
    print('|' + contents + '|')
    print(border)

box_string('Hello')
box_string('Python')
````

````python
# Example 5 — returning a boolean
# is_even(n) returns True if n is even, False otherwise.

def is_even(n):
    return n % 2 == 0

print(is_even(4))   # True
print(is_even(7))   # False
print(is_even(0))   # True
````

````python
# Example 6 — calculate() rewritten with nonlocal
# Each inner function writes to the enclosing 'result' variable
# instead of returning a value directly.

def calculate(op, x, y):
    result = None

    def add():
        nonlocal result
        result = x + y

    def subtract():
        nonlocal result
        result = x - y

    def multiply():
        nonlocal result
        result = x * y

    def divide():
        nonlocal result
        result = x / y

    if   op == 'add':      add()
    elif op == 'subtract': subtract()
    elif op == 'multiply': multiply()
    else:                  divide()

    return result

print(calculate('add', 10, 5))       # 15
print(calculate('subtract', 10, 5))  # 5
print(calculate('multiply', 10, 5))  # 50
print(calculate('divide', 10, 5))    # 2.0
````

````python
# Example 7 — tracing return values
# mystery(n) returns True when n is even, False when odd.
# The if/else branches both end with return, so the function
# always produces a boolean.

def mystery(n):
    if n % 2 == 0:
        return True
    else:
        return False

print(mystery(4))   # True  — 4 % 2 == 0
print(mystery(7))   # False — 7 % 2 == 1
````

````python
# Example 8 — tracing parameter passing
# main() calls mystery_avg(5, 7).
# Inside mystery_avg: z = 5 + 7 = 12, then z = 12 / 2.0 = 6.0
# The return value 6.0 is passed directly to print().

def main():
    a = 5
    b = 7
    print(mystery_avg(a, b))   # prints 6.0

def mystery_avg(x, y):
    z = x + y
    z = z / 2.0
    return z

main()
````
