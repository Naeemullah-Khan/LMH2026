# Part 1 — List Comprehensions

## 1.1 The Core Idea

A **list comprehension** is a compact way to build a list from an existing iterable.  
It replaces the classic "empty list + loop + append" pattern with a single readable expression.

```
[  expression   for item in iterable   if condition  ]
  ────────────  ─────────────────────  ─────────────
   what to         where to get            optional
   produce          the items              filter
```

All three parts live inside square brackets `[ ]`, which signals that the result is a **list**.

````python
# Example 1 — the problem comprehensions solve
# Classic approach: empty list, loop, append
squares_classic = []
for n in range(1, 6):
    squares_classic.append(n ** 2)
print(squares_classic)   # [1, 4, 9, 16, 25]

# Comprehension: one line, same result
squares = [n ** 2 for n in range(1, 6)]
print(squares)           # [1, 4, 9, 16, 25]
````

## 1.2 Transforming Values

The **expression** part can be any Python expression — arithmetic, method calls, function calls, f-strings.
Every element in the iterable is transformed by that expression.

````python
# Example 2 — various transformation expressions
words = ['python', 'data', 'science', 'ai']

upper_words  = [w.upper() for w in words]
lengths      = [len(w) for w in words]
labels       = [f'topic: {w}' for w in words]

print(upper_words)
print(lengths)
print(labels)
````

````python
# Example 3 — transforming numbers
temperatures_c = [0, 20, 37, 100]

# Celsius to Fahrenheit: F = C * 9/5 + 32
temperatures_f = [round(c * 9/5 + 32, 1) for c in temperatures_c]
print(temperatures_f)   # [32.0, 68.0, 98.6, 212.0]
````

## 1.3 Filtering with `if`

An optional `if condition` at the **end** of the comprehension acts as a filter:  
only items for which the condition is `True` are included.

````python
# Example 4 — simple filter
numbers = range(1, 21)

evens   = [n for n in numbers if n % 2 == 0]
odds    = [n for n in numbers if n % 2 != 0]
big     = [n for n in numbers if n > 15]

print(evens)
print(odds)
print(big)
````

````python
# Example 5 — filtering strings
words = ['apple', '', 'banana', '  ', 'cherry', '', 'date']

# Keep only non-empty, non-whitespace strings
clean = [w for w in words if w.strip()]
print(clean)   # ['apple', 'banana', 'cherry', 'date']
````

````python
# Example 6 — filtering a list of numbers by multiple conditions
numbers = range(-10, 11)

# Positive AND even
pos_even = [n for n in numbers if n > 0 and n % 2 == 0]
print(pos_even)

# Divisible by 3 OR by 5 (but not zero)
fizzbuzz_candidates = [n for n in numbers if n != 0 and (n % 3 == 0 or n % 5 == 0)]
print(fizzbuzz_candidates)
````

## 1.4 Transforming AND Filtering at the Same Time

Both the transformation expression and the `if` filter can appear together.  
The `if` is applied **before** the expression — only matching items get transformed.

````python
# Example 7 — transform and filter in one step
scores = [72, 95, 48, 88, 61, 100, 55]

# Double the score only for students who passed
boosted_passing = [s * 1.1 for s in scores if s >= 60]
print([round(v, 1) for v in boosted_passing])

# Clean and uppercase only non-empty names
raw_names = ['  aisha ', '', 'omar', '  ', '  layla ']
clean = [name.strip().title() for name in raw_names if name.strip()]
print(clean)
````

## 1.5 Conditional Expression Inside a Comprehension

A **conditional (ternary) expression** — `A if condition else B` — lives in the  
*transformation* part (before `for`), not after it.  
This transforms every item, choosing between two different outputs.

````python
# Example 8 — ternary expression in the transformation part
scores = [85, 42, 90, 55, 78, 30, 95]

# Label every score — no item is filtered out
labels = ['pass' if s >= 60 else 'fail' for s in scores]
print(labels)

# ⚠️  Compare the two placements:
# [x * 2 for x in data if x > 0]         ← filter: only positive items kept
# [x * 2 if x > 0 else 0 for x in data]  ← transform: ALL items kept, different output
````

````python
# Example 9 — both a ternary expression AND a filter
numbers = range(-5, 6)

# Keep only non-zero numbers, then label them positive or negative
result = ['pos' if n > 0 else 'neg' for n in numbers if n != 0]
print(result)
````

## 1.6 Nested List Comprehensions

You can place one `for` clause inside another to iterate over nested structures.  
The order matches how you would write the equivalent nested loops: **outer loop first, inner loop second**.

````python
# Example 10 — flattening a matrix
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Equivalent nested loops:
flat_loop = []
for row in matrix:
    for cell in row:
        flat_loop.append(cell)

# Nested comprehension:
flat = [cell for row in matrix for cell in row]

print(flat_loop)
print(flat)
print(flat_loop == flat)   # True
````

````python
# Example 11 — generating all (x, y) coordinate pairs
xs = [1, 2, 3]
ys = [10, 20]

pairs = [(x, y) for x in xs for y in ys]
print(pairs)
# [(1, 10), (1, 20), (2, 10), (2, 20), (3, 10), (3, 20)]
````

````python
# Example 12 — nested comprehension with a condition on the inner loop
# All (x, y) pairs where x < y, drawn from the same range
pairs = [(x, y) for x in range(1, 5) for y in range(1, 5) if x < y]
print(pairs)
````

# Part 2 — Dictionary Comprehensions

## 2.1 Syntax

A **dictionary comprehension** builds a `dict` from an iterable.  
It uses **curly braces** and requires both a key expression and a value expression.

```
{  key_expr : value_expr   for item in iterable   if condition  }
  ─────────   ──────────   ─────────────────────  ─────────────
   key part   value part       data source          optional
```

````python
# Example 13 — basic dictionary comprehension
# Number → its square
squares = {n: n ** 2 for n in range(1, 6)}
print(squares)

# Word → its length
words   = ['AI', 'Python', 'data', 'science']
lengths = {word: len(word) for word in words}
print(lengths)
````

````python
# Example 14 — building from two parallel sequences
keys   = ['name', 'gpa', 'year']
values = ['Aisha', 3.9, 2]

student = {k: v for k, v in zip(keys, values)}
print(student)
````

````python
# Example 15 — transforming an existing dictionary
prices = {'apple': 3.50, 'banana': 1.20, 'cherry': 6.80}

# Apply a 10 % discount to every price
discounted = {item: round(price * 0.90, 2) for item, price in prices.items()}
print(discounted)

# Convert all keys to uppercase
upper_keys = {k.upper(): v for k, v in prices.items()}
print(upper_keys)
````

````python
# Example 16 — filtering a dictionary
stock = {'apple': 0, 'banana': 15, 'cherry': 0, 'date': 8, 'elderberry': 3}

# Keep only items that are in stock
in_stock = {item: qty for item, qty in stock.items() if qty > 0}
print(in_stock)

# Keep only items with qty > 5
well_stocked = {item: qty for item, qty in stock.items() if qty > 5}
print(well_stocked)
````

````python
# Example 17 — inverting a dictionary (swap keys and values)
# Works correctly when values are unique
country_code = {'Saudi Arabia': 966, 'UAE': 971, 'Egypt': 20, 'Jordan': 962}

code_country = {code: name for name, code in country_code.items()}
print(code_country)
````

````python
# Example 18 — nested dict comprehension
# Build a multiplication table as a dict of dicts
table = {i: {j: i * j for j in range(1, 4)} for i in range(1, 4)}

for row, cols in table.items():
    print(f'{row}: {cols}')
````

# Part 3 — Set Comprehensions

## 3.1 Syntax

A **set comprehension** uses curly braces like a dict comprehension,  
but has **only one expression** (no colon).  
The result is a `set` — duplicates are automatically removed.

```
{  expression   for item in iterable   if condition  }
```

**Distinguishing `{}` syntax:**

| Syntax | Result |
|---|---|
| `{expr for x in it}` | set |
| `{k: v for x in it}` | dict |
| `{}` | **empty dict** (not a set!) |

````python
# Example 19 — basic set comprehension
words = ['banana', 'apple', 'cherry', 'apricot', 'blueberry', 'avocado']

# Unique first letters
first_letters = {word[0] for word in words}
print(first_letters)   # {'b', 'a', 'c'} — order not guaranteed

# Unique word lengths
lengths = {len(word) for word in words}
print(lengths)
````

````python
# Example 20 — deduplication via set comprehension
records = [
    'AI', 'Python', 'ai', 'DATA', 'python', 'Data', 'AI'
]

# Unique lowercase versions — duplicates are collapsed
unique = {word.lower() for word in records}
print(unique)
````

````python
# Example 21 — set comprehension with a condition
numbers = range(-10, 11)

# Set of absolute values of all negative numbers
neg_abs = {abs(n) for n in numbers if n < 0}
print(sorted(neg_abs))

# Set of all letters used in a sentence (ignoring spaces and punctuation)
sentence = 'The quick brown fox jumps over the lazy dog'
used_letters = {ch.lower() for ch in sentence if ch.isalpha()}
print(sorted(used_letters))   # all 26 letters of the alphabet
````

````python
# Example 22 — practical use: finding common elements
course_python = ['Aisha', 'Omar', 'Layla', 'Nour', 'Tariq']
course_ai     = ['Omar', 'Sana', 'Layla', 'Khalid', 'Aisha']

# Students in BOTH courses — using set comprehensions
set_python = {name.lower() for name in course_python}
set_ai     = {name.lower() for name in course_ai}

both = set_python & set_ai
print('In both courses:', both)

only_python = set_python - set_ai
print('Python only:    ', only_python)
````

# Part 4 — "Tuple Comprehension": Why It Does Not Exist

## 4.1 The Parenthesis Surprise

Python has list `[ ]`, dict `{ : }`, and set `{ }` comprehensions.  
You might expect `( )` to create a tuple comprehension — but it does **not**.  

Instead, `(expression for item in iterable)` creates a **generator expression**  
(covered in Part 5). This is intentional: tuples are for fixed-length records,  
while generators are for lazy, one-at-a-time iteration.

To produce a `tuple` from a comprehension-like expression, wrap a generator expression with `tuple()`.

````python
# Example 23 — parentheses create a generator, NOT a tuple
gen = (n ** 2 for n in range(1, 6))

print(type(gen))   # <class 'generator'>  — not a tuple!
print(gen)         # <generator object ...>
````

````python
# Example 24 — creating a tuple from a comprehension-like expression
# Wrap a generator expression in tuple()
squares_tuple = tuple(n ** 2 for n in range(1, 6))

print(type(squares_tuple))   # <class 'tuple'>
print(squares_tuple)         # (1, 4, 9, 16, 25)
````

````python
# Example 25 — when would you want this?
# Tuples are useful as immutable, hashable sequences —
# for example, as dictionary keys or set members.

words = ['apple', 'banana', 'cherry']

# A tuple of lengths (immutable)
lengths = tuple(len(w) for w in words)
print(lengths)         # (5, 6, 6)

# A tuple of (word, length) pairs
pairs = tuple((w, len(w)) for w in words)
print(pairs)           # (('apple', 5), ('banana', 6), ('cherry', 6))

# This tuple can now be used as a dict key or stored in a set
lookup = {pairs: 'my frozen record'}
print(lookup[pairs])
````

````python
# Example 26 — comparison of all four forms
source = range(1, 6)

as_list  = [n ** 2 for n in source]           # list  — [ ]
as_dict  = {n: n ** 2 for n in source}        # dict  — { : }
as_set   = {n ** 2 for n in source}           # set   — { }
as_gen   = (n ** 2 for n in source)           # generator — ( )
as_tuple = tuple(n ** 2 for n in source)      # tuple — tuple( )

print('list: ', as_list)
print('dict: ', as_dict)
print('set:  ', as_set)
print('gen:  ', as_gen)       # shows the object, not the values
print('tuple:', as_tuple)
````

# Part 5 — Generator Expressions

## 5.1 Lazy Evaluation

A **generator expression** looks exactly like a list comprehension but uses `( )` instead of `[ ]`.  
The key difference is **when** values are computed:

| | List comprehension | Generator expression |
|---|---|---|
| Evaluates values | All at once, immediately | One at a time, on demand |
| Memory | Stores the entire list | Stores almost nothing |
| Reusable | Yes — iterate multiple times | No — exhausted after one pass |
| Use when | You need the full list | You only iterate once, or data is large |

````python
# Example 27 — memory comparison
import sys

n = 1_000_000

list_comp = [x ** 2 for x in range(n)]
gen_expr  = (x ** 2 for x in range(n))

print(f'list size:      {sys.getsizeof(list_comp):>12,} bytes')
print(f'generator size: {sys.getsizeof(gen_expr):>12,} bytes')

# Both give the same sum
print('sum (list):', sum(list_comp))
print('sum (gen): ', sum(x ** 2 for x in range(n)))
````

````python
# Example 28 — generators are lazy: values computed on demand
def verbose_square(n):
    print(f'  computing {n}^2')
    return n ** 2

print('--- list comprehension: all computed immediately ---')
squares_list = [verbose_square(n) for n in range(1, 4)]
print('List ready:', squares_list)

print()
print('--- generator expression: nothing computed yet ---')
squares_gen = (verbose_square(n) for n in range(1, 4))
print('Generator created (no computation yet)')

print()
print('--- now consuming the generator one value at a time ---')
print(next(squares_gen))   # computed only when requested
print(next(squares_gen))
print(next(squares_gen))
````

## 5.2 Using Generators with Built-in Functions

Built-in functions like `sum()`, `max()`, `min()`, `any()`, and `all()` consume iterables one item at a time.  
Feeding them a generator expression avoids building an intermediate list.

````python
# Example 29 — generators in built-in functions
data = [3, -1, 7, -4, 2, -9, 6]

# sum of squares of all positive numbers
total = sum(x ** 2 for x in data if x > 0)
print('sum of squares of positives:', total)

# largest absolute value
largest = max(abs(x) for x in data)
print('largest absolute value:', largest)

# True if ANY value is negative
any_neg = any(x < 0 for x in data)
print('any negative?', any_neg)

# True only if ALL values are positive
all_pos = all(x > 0 for x in data)
print('all positive?', all_pos)
````

````python
# Example 30 — short-circuit evaluation with generators
# any() and all() stop early when the answer is known.
# This is ONLY possible with a generator (a list is fully built first).

def check(n):
    print(f'  checking {n}')
    return n > 0

numbers = [-5, -2, 3, -1, 8]

print('any() stops at the first True:')
result = any(check(n) for n in numbers)   # stops when it finds 3
print('result:', result)
````

````python
# Example 31 — generators are exhausted after one pass
gen = (n ** 2 for n in range(1, 4))

print('First pass:',  list(gen))   # [1, 4, 9]
print('Second pass:', list(gen))   # []  — generator is exhausted!

# To iterate multiple times, use a list comprehension or recreate the generator
squares = [n ** 2 for n in range(1, 4)]
print('List, pass 1:', squares)
print('List, pass 2:', squares)   # still [1, 4, 9]
````

## 5.3 Generator Expressions vs. List Comprehensions — Decision Guide

| Use a **generator expression** when… | Use a **list comprehension** when… |
|---|---|
| You only need to iterate **once** | You need to iterate **multiple times** |
| The data set is **large** | The collection is **small** |
| You are feeding directly into `sum`, `max`, `any`, `all` | You need **indexing** `result[i]` |
| You want to **save memory** | You need to **store** the results for later use |

# Part 6 — Comprehensions vs. Explicit Loops

## 6.1 The Readability Principle

> **Readable code is better than compressed code.**

Comprehensions shine when the logic is a single, clear transformation or filter.  
An explicit `for` loop is better when the body needs several steps, debug prints, or complex branching.

| Prefer a comprehension when… | Prefer an explicit loop when… |
|---|---|
| Logic is one expression | Logic needs multiple lines or variables |
| Condition is simple | You need intermediate debug prints |
| You are building a new collection | You are changing existing state |
| It reads like natural language | A colleague would need to pause and decode it |

````python
# Example 32 — comprehension wins: clean, single-step transformation
temperatures_c = [0, 20, 37, 100]

# ✅ Comprehension — one clear formula
fahrenheit = [c * 9/5 + 32 for c in temperatures_c]

# ❌ Loop — unnecessarily verbose for this task
fahrenheit_loop = []
for c in temperatures_c:
    fahrenheit_loop.append(c * 9/5 + 32)

print(fahrenheit)
print(fahrenheit == fahrenheit_loop)
````

````python
# Example 33 — explicit loop wins: multi-step logic with side effects
raw_lines = ['  42abc ', '  17  ', '  not_a_number ', '  99  ']

# ❌ Forcing this into a comprehension is hard to read
# results = [int(line.strip()) for line in raw_lines if line.strip().isdigit()]

# ✅ Explicit loop — room to handle each step clearly
results = []
for line in raw_lines:
    stripped = line.strip()
    if not stripped.isdigit():
        print(f'Skipping non-numeric value: {repr(stripped)}')
        continue
    results.append(int(stripped))

print('Valid numbers:', results)
````

````python
# Example 34 — anti-pattern: comprehension used purely for side effects
names = ['aisha', 'omar', 'layla']

# ❌ Using a comprehension just to call print — produces a useless list of None
# [print(name.title()) for name in names]

# ✅ Correct: use a loop when the purpose is a side effect, not building a list
for name in names:
    print(name.title())
````

````python
# Example 35 — anti-pattern: modifying a list while iterating it
numbers = [1, 2, 3, 4, 5, 6]

# ❌ Dangerous: removing from a list while iterating causes items to be skipped
# for n in numbers:
#     if n % 2 == 0:
#         numbers.remove(n)

# ✅ Safe: build a new list with a comprehension
odds_only = [n for n in numbers if n % 2 != 0]
print(odds_only)
````

# Part 7 — Nested Comprehensions and Refactoring for Clarity

## 7.1 When Nesting Is Acceptable

A nested comprehension is readable when:
- The structure is a simple 2-D iteration (e.g., flatten a list of lists)
- Both loops are short and clearly named
- The expression is simple (no extra conditions on both loops)

It becomes a problem when a reader has to mentally re-order the clauses to understand it.

````python
# Example 36 — acceptable nesting: flatten a list of lists
paragraphs = [
    ['Python', 'is', 'great'],
    ['AI', 'is', 'transforming', 'everything'],
    ['Data', 'science', 'matters']
]

# Readable: outer=paragraph, inner=word
all_words = [word for paragraph in paragraphs for word in paragraph]
print(all_words)
````

````python
# Example 37 — acceptable nesting: Cartesian product
suits  = ['♠', '♥', '♦', '♣']
values = ['A', '2', '3', 'J', 'Q', 'K']

deck = [f'{v}{s}' for s in suits for v in values]
print(f'{len(deck)} cards — first 8: {deck[:8]}')
````

## 7.2 When Nesting Needs Refactoring

The warning signs that a nested comprehension is too complex:

- More than two `for` clauses
- Conditions on both the outer and inner loop
- A reader needs to re-read it twice to understand it

**Three refactoring strategies:**
1. Break it into an explicit nested loop
2. Extract the inner comprehension into a helper variable or function
3. Use a helper function that makes the intent explicit

````python
# Example 38 — overly complex nested comprehension
matrix = [
    [1, -2,  3],
    [-4, 5, -6],
    [7, -8,  9]
]

# ❌ Hard to read: nested comprehension with conditions on both levels
result_hard = [cell * 2 for row in matrix
               if sum(row) > 0
               for cell in row
               if cell > 0]
print('Hard version:', result_hard)

# ✅ Strategy 1: explicit nested loops with clear variable names
result_clear = []
for row in matrix:
    if sum(row) > 0:            # only process rows with positive sum
        for cell in row:
            if cell > 0:        # only keep positive cells
                result_clear.append(cell * 2)

print('Clear version:', result_clear)
print('Same result:  ', result_hard == result_clear)
````

````python
# Example 39 — Strategy 2: break into intermediate variables
matrix = [
    [1, -2,  3],
    [-4, 5, -6],
    [7, -8,  9]
]

# Step 1: filter to rows with positive sum
positive_rows = [row for row in matrix if sum(row) > 0]

# Step 2: flatten and keep only positive cells, then double them
result = [cell * 2 for row in positive_rows for cell in row if cell > 0]

print('Positive rows:', positive_rows)
print('Result:       ', result)
````

````python
# Example 40 — Strategy 3: extract into a helper function
def process_row(row):
    """Return doubled positive cells, or empty list if row sum is not positive."""
    if sum(row) <= 0:
        return []
    return [cell * 2 for cell in row if cell > 0]

matrix = [
    [1, -2,  3],
    [-4, 5, -6],
    [7, -8,  9]
]

# The top-level comprehension is now trivially readable
result = [value for row in matrix for value in process_row(row)]
print(result)
````

## 7.3 Comprehension Performance Notes

Comprehensions are generally **faster** than equivalent `for` loops with `append` because:
- The loop runs at the C level inside the CPython interpreter
- There is no repeated method lookup for `.append`

However, this speed difference is rarely significant.  
**Never sacrifice readability purely for performance unless profiling tells you to.**

````python
# Example 41 — timing comparison: comprehension vs loop
import timeit

setup = 'data = range(10_000)'

t_comp = timeit.timeit(
    '[x ** 2 for x in data]',
    setup=setup, number=1000
)
t_loop = timeit.timeit(
    'result = []\nfor x in data: result.append(x**2)',
    setup=setup, number=1000
)

print(f'Comprehension: {t_comp:.3f}s')
print(f'Explicit loop: {t_loop:.3f}s')
print(f'Comprehension is ~{t_loop/t_comp:.1f}x faster')
````

# Summary

| Concept | Key Points |
|---|---|
| **List Comprehensions** | `[expression for item in iterable if condition]` — compact way to build lists; allows transformation and filtering. |
| **Dictionary Comprehensions** | `{key: value for item in iterable if condition}` — builds dictionaries; requires key and value expressions. |
| **Set Comprehensions** | `{expression for item in iterable if condition}` — builds sets; removes duplicates; only one expression. |
| **Tuple Comprehensions** | Do not exist. `(expression for ...)` creates a **generator expression**, not a tuple. Use `tuple(...)` to convert a generator to a tuple. |
| **Generator Expressions** | `(expression for item in iterable)` — lazy evaluation, values computed on demand; memory efficient; iterable only once. |
| **Comprehensions vs. Loops** | Prefer comprehensions for concise transformations/filters; use explicit loops for complex logic, side effects, or debugging. |
| **Nested Comprehensions** | Readable for simple 2-D iterations. Refactor if too complex (multiple conditions, many `for` clauses) using explicit loops or helper functions. |
| **Performance** | Comprehensions are generally faster than `for` loops for simple tasks, but readability should be prioritized over minor speed gains.
