# What is a Notebook?

A notebook contains cells, which are distinct blocks of content. There are two primary types of cells:

*   **Code cells:** These contain executable code (like Python) and display their output directly below them.
*   **Markdown cells:** These contain formatted text, explanations, equations, images, and notes. They are rendered to display rich content.

## Why Notebooks are Common in AI/ML

Notebooks are widely used in AI/ML because they combine various aspects of a project—experimentation, explanation, visualization, and results—all in one interactive document. This makes them ideal for iterative development, sharing, and presenting work.

## Caution: Execution Order Matters

It's crucial to understand that the order in which cells are executed in a notebook can significantly impact the results. Running cells out of order, especially for beginners, can lead to confusion or errors because variables and states might not be as expected.

## Beyond Static Code: The Interactive Advantage

Unlike traditional script files, notebooks offer a highly interactive and dynamic environment. You don't just write code and run it all at once; you execute it in smaller, manageable chunks (cells).

This allows for:

*   **Incremental Development:** Build and test your code step by step.
*   **Immediate Feedback:** See the output of each code block right away.
*   **Easy Experimentation:** Change a piece of code, re-run just that cell, and observe the new results without restarting everything.

## Key Benefits of Notebooks for Learning and AI/ML

Notebooks excel in educational and AI/ML contexts because they:

*   **Combine Narrative and Code:** Explain your thoughts, show your code, and display results, all in one document.
*   **Facilitate Data Exploration:** Quickly load data, visualize it, and perform transformations with instant results.
*   **Simplify Iteration:** Rapidly prototype and refine models or algorithms.
*   **Enhance Collaboration:** Share your entire thought process, code, and findings with others in an easy-to-understand format.

## How to Interact with This Notebook

To get the most out of this learning experience, follow these steps:

1.  **Select a Cell:** Click anywhere on a cell (like this one!) to select it.
2.  **Run a Cell:** With a cell selected, press `Shift + Enter` (or click the 'play' button next to the cell) to execute its content.
    *   For **Markdown cells**, running them renders the formatted text.
    *   For **Code cells**, running them executes the Python code and displays any output below the cell.
3.  **Follow Sequentially:** It's generally best to run cells from top to bottom, in order, to ensure the notebook's state (e.g., variables, functions) is consistent.

Feel free to experiment! You can modify the code in a code cell and re-run it to see what happens. This is encouraged!

## Your First Python Program

Let's start with a classic: printing a message to the console.

````python
print("Hello, World!")
````

### What happens?

1.  Python reads the instruction `print("Hello, World!")`.
2.  It recognizes `print` as a built-in function, which means it performs a specific task.
3.  It sends the text `"Hello, World!"` to the screen, which is displayed as output below the cell.

### The `print()` function

The `print()` function is used to display output to the console. It can display various types of data.

````python
print("Python is fun")
print(42)
print("The answer is", 42)
````

### Vocabulary

*   `print`: This is the function name. Functions are reusable blocks of code that perform a specific action.
*   Parentheses `()`: These hold the information (arguments) that you want to give to the function.
*   `"Python is fun"`: This is a **string**, which means text data. Strings are always enclosed in quotes (single or double).
*   `42`: This is a **number** (specifically, an integer).

### `print()` with `sep` (separator)

By default, `print()` separates multiple values with a space. You can change this behavior using the `sep` argument.

````python
print("A", "B", "C")
print("A", "B", "C", sep="-")
print("2026", "05", "03", sep="/")
````

### `print()` with `end`

By default, `print()` adds a new line character at the end of its output. You can change what it ends with using the `end` argument.

````python
print("Hello")
print("World")

# Change the ending:
print("Hello", end=" ")
print("World")
````

## Comments

Comments are notes within your code that are meant for human readers. Python ignores them when it runs the program. Comments start with a `#` symbol.

````python
# This program prints a greeting
print("Welcome to Python") # Display message to user
````

Good comments explain *why* something is done, not just *what* the code already clearly says.

*   **Weak comment:**
    ```python
    print("Hello") # prints Hello
    ```
*   **Better comment:**
    ```python
    # First output confirms that the environment is working
    print("Hello")
    ```

## Docstrings

A **docstring** is a special string used to document a module, function, class, or method. It provides a concise summary of what the code does.

YouYou will understand functions more deeply later, but here is an early preview:

````python
def greet():
  """Display a simple greeting message."""
  print("Hello!")

greet()
````

### Beginner takeaway

Comments and docstrings are crucial for making code understandable, not just for others, but also for your future self. Good programmers write code that is clear for both computers and humans.

## Python Syntax: Tiny Details Matter

Computers are very strict. Even small syntax details can drastically change the meaning of your code or cause errors. Python will tell you when you have a syntax error.

*   **Correct:**
    ```python
    print("Hello")
    ```
*   **Incorrect (missing closing parenthesis):**
    ```python
    print("Hello"
    ```
    (If you try to run this, you will get a `SyntaxError`.)


**Syntax** is the grammar of a programming language—the rules for how statements must be structured. **Logic** is the meaning of the steps—what the program is actually trying to accomplish. Both are equally important for writing functional and understandable code.
