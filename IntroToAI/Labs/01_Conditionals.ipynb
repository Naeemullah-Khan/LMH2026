# Part 1 — `if`, `elif`, `else`: Syntax and Structure

## 1.1 The Basic `if` Statement

A program often needs to make a **decision** — do one thing under certain conditions, something else otherwise.  
Python uses the `if` statement to express decisions.

### Syntax

```
if <condition>:
    <statement(s) executed when condition is True>
```

Key rules:
- The condition is any expression that evaluates to `True` or `False`.
- The **colon** (`:`) at the end of the `if` line is **required**.
- The body must be **indented** (4 spaces is the Python convention).

If there is nothing to do when the condition is `False`, the `else` branch can be omitted entirely.

````python
# Example 1 — if only (no else needed)
temperature = 35   # degrees Celsius

if temperature > 30:
    print("It's a hot day — stay hydrated!")

print("Program continues here regardless.")
````

## 1.2 Adding `else`

When you need to do something **either way**, add an `else` branch:

```
if <condition>:
    <statements when True>
else:
    <statements when False>
```

Exactly **one** of the two branches will execute — never both, never neither.

````python
# Example 2 — if / else: pass or fail
score = 72

if score >= 60:
    print("Result: PASS")
else:
    print("Result: FAIL")
````

## 1.3 Multiple Alternatives with `elif`

When there are **more than two** possible outcomes, chain conditions with `elif`  
(`elif` is short for "else if"):

```
if <condition1>:
    <block1>
elif <condition2>:
    <block2>
elif <condition3>:
    <block3>
else:
    <default block>
```

Python tests conditions **in order** and executes the **first** branch whose condition is `True`.  
All remaining branches are skipped, even if their conditions would also be `True`.

````python
# Example 3 — elif chain: letter grade calculator
score = float(input("Enter your score (0–100): "))

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Grade: {grade}")
````

### ⚠️ Order Matters!

Always test the **most restrictive** condition first.  
Compare the two versions below — they look similar but behave very differently:

````python
# Example 4 — wrong vs correct ordering
score = 95

# ❌ WRONG ORDER — the first branch catches everything >= 60
print("--- Wrong order ---")
if score >= 60:
    print("D")          # always prints for any passing score!
elif score >= 70:
    print("C")          # never reached
elif score >= 80:
    print("B")          # never reached
elif score >= 90:
    print("A")          # never reached

# ✅ CORRECT ORDER — most restrictive first
print("--- Correct order ---")
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
````

````python
# Example 5 — season classifier
# Determine the season in Saudi Arabia based on a month number
month = int(input("Enter a month number (1–12): "))

if month == 12 or month == 1 or month == 2:
    season = "Winter"
elif month == 3 or month == 4 or month == 5:
    season = "Spring"
elif month == 6 or month == 7 or month == 8 or month == 9:
    season = "Summer"
elif month == 10 or month == 11:
    season = "Autumn"
else:
    season = "Invalid month"

print(f"Month {month} → {season}")
````

# Part 2 — Nested Conditionals and Readability

## 2.1 What is a Nested Conditional?

A **nested conditional** is an `if` statement placed **inside** another `if` statement.  
This is used when a decision only makes sense **after** another decision has already been made.

Think of it like a two-stage gate: you can only reach the second gate if you already passed the first.

````python
# Example 6 — nested conditional: ATM withdrawal
# To dispense money, TWO conditions must hold in sequence:
#   1. The PIN must be correct.
#   2. The requested amount must not exceed the balance.

correct_pin = 4821
balance     = 1500.0

pin_entered = int(input("Enter your PIN: "))
amount      = float(input("Amount to withdraw: "))

if pin_entered == correct_pin:
    # We are inside the 'correct PIN' branch
    if amount <= balance:
        balance -= amount
        print(f"Dispensing SAR {amount:.2f}. Remaining balance: SAR {balance:.2f}")
    else:
        print("Insufficient funds.")
else:
    print("Incorrect PIN. Access denied.")
````

## 2.2 Readability Considerations

Nesting is sometimes the right tool, but **too many levels of nesting make code hard to read**.  
When conditions are **independent of each other**, a flat `elif` chain is often clearer.

Compare the two versions below — they produce identical output:

````python
# Example 7 — nested vs flat: classifying a number
x = 7

# ❌ Deeply nested — hard to follow
print("--- Nested ---")
if x > 0:
    if x % 2 == 0:
        print("Positive and even")
    else:
        print("Positive and odd")
else:
    if x % 2 == 0:
        print("Non-positive and even")
    else:
        print("Non-positive and odd")

# ✅ Flat elif chain — easier to read
print("--- Flat ---")
if x > 0 and x % 2 == 0:
    print("Positive and even")
elif x > 0 and x % 2 != 0:
    print("Positive and odd")
elif x % 2 == 0:
    print("Non-positive and even")
else:
    print("Non-positive and odd")
````

````python
# Example 8 — theme park ticket pricing
# Pricing depends on age first, then student status within each age group.
# Nesting is appropriate here because the inner decision depends on the outer one.

age        = int(input("Enter visitor age: "))
is_student = input("Are you a student? (yes/no): ").lower() == "yes"

if age < 5:
    price = 0
elif age <= 17:
    if is_student:
        price = 20
    else:
        price = 30
else:
    if is_student:
        price = 40
    else:
        price = 60

print(f"Ticket price: SAR {price}")
````

# Part 3 — Truthiness and Falsiness

## 3.1 Every Value Has a Boolean Interpretation

In Python, **any** value can be used as a condition — not just `True` and `False`.  
Python assigns every value a **truthiness**: it is either *truthy* (behaves like `True`) or *falsy* (behaves like `False`).

### Falsy Values

The values that Python treats as `False` in a condition are:

| Value | Type | Meaning |
|---|---|---|
| `False` | bool | The boolean false literal |
| `0` | int | The integer zero |
| `0.0` | float | The float zero |
| `""` | str | The empty string |
| `None` | NoneType | The absence of a value |

**Everything else** — any non-zero number, any non-empty string — is **truthy**.

You can inspect a value's truthiness directly with the built-in `bool()` function:

````python
# Example 9 — checking truthiness with bool()

# Falsy values
print(bool(False))   # False
print(bool(0))       # False
print(bool(0.0))     # False
print(bool(""))      # False  (empty string)
print(bool(None))    # False

print("---")

# Truthy values
print(bool(True))    # True
print(bool(1))       # True
print(bool(-99))     # True  (any non-zero number)
print(bool("hi"))    # True  (any non-empty string)
print(bool(" "))     # True  (a space character is NOT an empty string!)
````

## 3.2 Truthiness in Practice — Strings

The most common use of truthiness is checking whether a string **has content** without explicitly writing `!= ""`.

````python
# Example 10 — checking for a non-empty string
username = input("Enter your username (leave blank to cancel): ")

# Both versions do the same thing:

# Explicit version
if username != "":
    print("Explicit check: Welcome,", username)
else:
    print("Explicit check: No username entered.")

# Pythonic version using truthiness
if username:
    print("Truthiness check: Welcome,", username)
else:
    print("Truthiness check: No username entered.")
````

## 3.3 Truthiness in Practice — Numbers

A numeric variable is falsy only when it is exactly zero.  
This is useful for checking whether a counter or quantity has a meaningful value.

````python
# Example 11 — zero vs non-zero
attempts = int(input("How many login attempts? "))

if attempts:
    print(f"There have been {attempts} login attempt(s).")
else:
    print("No login attempts recorded yet.")
````

## 3.4 `None` — The Absence of a Value

`None` is a special constant that means **"no value"** or **"not applicable"**.  
It is commonly used as a placeholder when a variable has not yet been given a meaningful value.

`None` is **falsy**, so it can be used directly in a condition.  
However, it is clearer and safer to use `is None` / `is not None` for explicit checks.

````python
# Example 12 — None as a placeholder
# Before a user enters their name, the variable holds None
player_name = None

if player_name is None:
    print("No player registered yet.")
else:
    print(f"Current player: {player_name}")

# Simulate the user registering
player_name = input("Enter your player name: ")

if player_name is not None and player_name != "":
    print(f"Welcome, {player_name}!")
else:
    print("Registration cancelled.")
````

## ⚠️ Truthiness Pitfall — `0` Can Be a Valid Answer!

Be careful when `0` is a **meaningful result** and not simply "missing".  
Relying on falsiness in that case would treat a real answer of `0` as though nothing was available.

````python
# Example 13 — the zero pitfall
# A temperature sensor returns a number, or None if the sensor is offline.
reading = 0   # valid reading: exactly 0 degrees

# ❌ BUG — treats a real reading of 0 as "no reading"
if reading:
    print("❌ Sensor reading:", reading)
else:
    print("❌ No reading available.")   # incorrectly prints this!

print("---")

# ✅ Correct — explicitly check for None
if reading is not None:
    print("✅ Sensor reading:", reading)
else:
    print("✅ No reading available.")
````

# Part 4 — Ternary (Conditional) Expressions

## 4.1 What is a Ternary Expression?

A **ternary expression** (also called a **conditional expression**) lets you write a simple `if/else` decision **on a single line** — and it produces a **value** rather than executing a block.

### Syntax

```python
<value_if_true>   if <condition>   else   <value_if_false>
```

Think of it as answering: *"Give me X if the condition holds, otherwise give me Y."*

````python
# Example 14 — if/else block vs. ternary expression (same result)
age = 20

# Traditional if/else — 3 lines
if age >= 18:
    status = "adult"
else:
    status = "minor"
print("Block:   ", status)

# Ternary expression — 1 line
status = "adult" if age >= 18 else "minor"
print("Ternary: ", status)
````

## 4.2 Ternary Expressions — Common Use Cases

````python
# Example 15 — conditional assignment
balance    = 250.0
withdrawal = 300.0

result = "Approved" if withdrawal <= balance else "Declined"
print(f"Transaction: {result}")
````

````python
# Example 16 — ternary inside an f-string
# Correctly pluralise "item" vs "items" based on the count
num_items = 1

print(f"You have {num_items} item{'s' if num_items != 1 else ''} in your order.")
````

````python
# Example 17 — supplying a default when a variable is empty
nickname = ""   # user left the field blank

display_name = nickname if nickname else "Anonymous"
print(f"Hello, {display_name}!")
````

````python
# Example 18 — ternary directly inside a print call
temperature = 15

print("Bring a jacket." if temperature < 20 else "No jacket needed.")
````

## 4.3 When NOT to Use Ternary

Ternary is great for **simple, readable** one-liners.  
Avoid it when the condition or values are complex, or when you need to execute multiple statements.

````python
# Example 19 — nested ternary vs regular if/elif/else
x = 5
y = 10
z = 3

# ❌ Nested ternary — very hard to read
result = "x" if x > y else ("y" if y > z else "z")
print("Hard to read:", result)

# ✅ Regular if/elif/else — clear intent
if x > y and x > z:
    result = "x"
elif y > z:
    result = "y"
else:
    result = "z"
print("Easy to read:", result)
````

````python
# Example 20 — converting an if/else block into a ternary expression
gpa  = 3.7
hour = 9

# Original if/else:
# if gpa >= 3.5:
#     honour = "Dean's List"
# else:
#     honour = "Regular Standing"
#
# if hour < 12:
#     greeting = "Good morning"
# else:
#     greeting = "Good afternoon"

# Ternary versions:
honour   = "Dean's List" if gpa >= 3.5 else "Regular Standing"
greeting = "Good morning" if hour < 12 else "Good afternoon"

print(honour)
print(greeting)
````

# Summary

| Concept | Key Points |
|---|---|
| `if / elif / else` | Test conditions in order; first `True` branch wins; `else` is optional |
| Nested conditionals | Use when one decision depends on another; flatten with `elif` when conditions are independent |
| Falsy values | `False`, `0`, `0.0`, `""`, `None` all behave as `False` in a condition |
| `None` | Represents the absence of a value; prefer `is None` / `is not None` for explicit checks |
| Ternary expression | `val_if_true if condition else val_if_false` — ideal for simple one-liner assignments; avoid nesting them |
