# Part 1 — String Immutability and Internal Representation

## 1.1 What Is a String?

A **string** is an ordered sequence of characters. Python stores every string as a series of individual characters, each at a numbered position called an **index**.

```
 P  y  t  h  o  n
 0  1  2  3  4  5     ← positive indexes
-6 -5 -4 -3 -2 -1     ← negative indexes
```

Key facts:
- Indexes start at `0`, not `1`.
- Negative indexes count from the **end**: `-1` is the last character.
- `len()` returns the total number of characters.

````python
# Example 1 — indexing and length
word = "Python"

print(len(word))    # 6
print(word[0])      # P  — first character
print(word[5])      # n  — last character (index = length - 1)
print(word[-1])     # n  — same last character, using negative index
print(word[-6])     # P  — same first character, using negative index
````

## 1.2 Slicing — Extracting a Portion of a String

A **slice** extracts a substring using the syntax `string[start:stop]`.  
The result includes index `start` but **stops before** index `stop`.

An optional third value controls the **step**: `string[start:stop:step]`.

````python
# Example 2 — slicing
course = "Intermediate Python"

print(course[0:12])    # Intermediate   (indexes 0 to 11)
print(course[13:])     # Python         (from index 13 to the end)
print(course[:12])     # Intermediate   (from the start up to index 11)
print(course[-6:])     # Python         (last 6 characters)
print(course[::2])     # every second character
print(course[::-1])    # reversed string
````

## 1.3 Strings Are Immutable

**Immutable** means a string cannot be changed after it is created.  
You cannot replace, insert, or delete individual characters in place.

Instead, you always **create a new string** that incorporates the change.

````python
# Example 3 — immutability: you cannot change a character in place
word = "Python"

# This raises a TypeError:
# word[0] = "J"   # ← TypeError: 'str' object does not support item assignment

# Instead, build a new string:
new_word = "J" + word[1:]   # "J" + "ython"
print(new_word)   # Jython
print(word)       # Python  — the original is untouched
````

````python
# Example 4 — string methods always return a NEW string
text = "  hello world  "

cleaned = text.strip()    # strip() returns a new string

print(repr(text))         # '  hello world  '  — original unchanged
print(repr(cleaned))      # 'hello world'      — new cleaned version
````

## 1.4 String Concatenation and Repetition

Since strings are immutable, `+` and `*` always produce **new** strings.  
They never modify the original.

````python
# Example 5 — concatenation and repetition
first = "Artificial"
second = "Intelligence"

full = first + " " + second
print(full)           # Artificial Intelligence

divider = "-" * 30
print(divider)        # ------------------------------

# += reassigns the variable to a new string object
greeting = "Hello"
greeting += ", world!"
print(greeting)       # Hello, world!
````

## 1.5 The `in` Operator — Membership Testing

The `in` operator checks whether a substring appears anywhere inside a string.

````python
# Example 6 — membership with in and not in
sentence = "Python is used in data science and AI"

print("data" in sentence)       # True
print("Java" in sentence)       # False
print("AI" not in sentence)     # False

# Practical use in a conditional:
filename = "report_2026.csv"
if ".csv" in filename:
    print("This is a CSV file.")
````

# Part 2 — String Methods

## 2.1 Case Methods: `upper()`, `lower()`, `title()`, `capitalize()`

These methods change the letter case of a string.  
They are often used to **normalise** user input before comparing or storing it.

````python
# Example 7 — case conversion methods
name = "aisha AL-RASHID"

print(name.upper())       # AISHA AL-RASHID
print(name.lower())       # aisha al-rashid
print(name.title())       # Aisha Al-Rashid   (first letter of each word capitalised)
print(name.capitalize())  # Aisha al-rashid   (only the very first letter capitalised)
````

````python
# Example 8 — why normalisation matters
# Case-sensitive comparison fails without normalisation
command = input("Type YES to continue: ")

# Without normalisation:
if command == "yes":
    print("Matched 'yes'")
elif command == "YES":
    print("Matched 'YES'")
else:
    print("No match — the user typed:", repr(command))

print("---")

# With normalisation — handles any capitalisation:
if command.lower() == "yes":
    print("Continuing...")
else:
    print("Cancelled.")
````

## 2.2 Whitespace Removal: `strip()`, `lstrip()`, `rstrip()`

User input often carries **accidental whitespace** at the start or end.  
These methods remove it without touching the content in the middle.

| Method | Removes whitespace from |
|---|---|
| `strip()` | Both ends |
| `lstrip()` | Left end only |
| `rstrip()` | Right end only |

````python
# Example 9 — strip variants
raw = "   machine learning   "

print(repr(raw.strip()))    # 'machine learning'
print(repr(raw.lstrip()))   # 'machine learning   '
print(repr(raw.rstrip()))   # '   machine learning'
````

````python
# Example 10 — stripping specific characters
# strip() can remove any characters you specify, not just spaces
messy = "###important###"
print(messy.strip("#"))    # important

path = "/home/user/data/"
print(path.strip("/"))     # home/user/data

# Clean a line read from a file (newline + spaces)
line = "   Ahmed   \n"
print(repr(line.strip()))  # 'Ahmed'
````

## 2.3 Splitting: `split()` and `splitlines()`

`split()` breaks a string into parts and returns them as a list.  
By default it splits on whitespace; you can specify any delimiter.

````python
# Example 11 — split() on whitespace and on a delimiter
sentence = "AI helps science and society"
words = sentence.split()
print(words)             # ['AI', 'helps', 'science', 'and', 'society']
print(len(words))        # 5 words

csv_row = "Aisha,95,AI,Year2"
fields = csv_row.split(",")
print(fields)            # ['Aisha', '95', 'AI', 'Year2']

# Limit the number of splits with maxsplit
log = "2026-05-08 ERROR disk full"
date, rest = log.split(" ", maxsplit=1)
print(date)   # 2026-05-08
print(rest)   # ERROR disk full
````

````python
# Example 12 — splitlines()
# Splits a multi-line string into individual lines
poem = "Roses are red\nViolets are blue\nPython is great\nAnd so are you"

lines = poem.splitlines()
print(lines)
print("Number of lines:", len(lines))
````

## 2.4 Joining: `join()`

`join()` is the **inverse** of `split()`. It combines a sequence of strings into one,  
placing a chosen **separator** between each piece.

The syntax feels unusual at first: the separator string **calls** `join()`.

````python
# Example 13 — join()
words = ["AI", "helps", "science"]

sentence = " ".join(words)       # space separator
print(sentence)                  # AI helps science

csv_line = ",".join(words)       # comma separator
print(csv_line)                  # AI,helps,science

path = "/".join(["home", "user", "data", "results.txt"])
print(path)                      # home/user/data/results.txt

# Join with no separator to concatenate tightly
chars = ["P", "y", "t", "h", "o", "n"]
print("".join(chars))            # Python
````

````python
# Example 14 — split then join: a common text-cleaning pattern
# Collapse multiple spaces into one by splitting on whitespace then joining
messy = "Python    is   very    powerful"
clean = " ".join(messy.split())
print(clean)   # Python is very powerful
````

## 2.5 Replacing: `replace()`

`replace(old, new)` returns a new string with **every occurrence** of `old` swapped for `new`.  
An optional third argument limits how many replacements are made.

````python
# Example 15 — replace()
text = "the cat sat on the mat and the cat"

# Replace all occurrences
print(text.replace("cat", "dog"))         # the dog sat on the mat and the dog

# Replace only the first occurrence
print(text.replace("cat", "dog", 1))      # the dog sat on the mat and the cat

# Replace produces a new string — original is unchanged
print(text)
````

````python
# Example 16 — replace() for cleaning text
user_input = "  my   phone: +966-512-345-678  "

# Remove dashes from a phone number
digits_only = user_input.strip().replace("-", "")
print(digits_only)   # my   phone: +966512345678
````

## 2.6 Searching: `find()`, `rfind()`, `index()`

`find(substring)` returns the **index** of the first occurrence of `substring`.  
If not found, it returns `-1` — it never raises an error.

`index()` works the same way but **raises a `ValueError`** if the substring is not found.  
`rfind()` searches from the **right** (last occurrence).

Both accept optional `start` and `end` arguments to limit the search range.

````python
# Example 17 — find() and rfind()
text = "Python is powerful. Python is readable."

print(text.find("Python"))     # 0   — first occurrence
print(text.rfind("Python"))    # 20  — last occurrence
print(text.find("Java"))       # -1  — not found; no error

# Search within a range: find from index 5 onward
print(text.find("Python", 5))  # 20  — skips the first occurrence
````

````python
# Example 18 — safe search with find() vs. crashing with index()
email = "student@university.edu"

# find() — safe: returns -1 if not found
at_pos = email.find("@")
if at_pos != -1:
    domain = email[at_pos + 1:]
    print("Domain:", domain)   # university.edu

# index() — raises ValueError if not found
try:
    pos = email.index("#")
except ValueError:
    print("'#' not found in email address")
````

## 2.7 Prefix and Suffix Checks: `startswith()`, `endswith()`

These return `True` or `False` and are more readable than slicing for checking the beginning or end of a string.

````python
# Example 19 — startswith() and endswith()
filename = "experiment_results_2026.csv"

print(filename.endswith(".csv"))           # True
print(filename.endswith(".xlsx"))          # False
print(filename.startswith("experiment"))   # True
print(filename.startswith("result"))       # False
````

````python
# Example 20 — passing a tuple to check multiple prefixes/suffixes at once
filename = "lecture_notes.pdf"

# Check if the file is any kind of document
if filename.endswith((".pdf", ".docx", ".txt")):
    print(f"'{filename}' is a text document.")

url = "https://www.university.edu"
if url.startswith(("http://", "https://")):
    print("Valid web URL")
````

## 2.8 Testing String Contents: `isdigit()`, `isalpha()`, `isspace()`

These inspection methods return `True` or `False` and are useful for validating input.

````python
# Example 21 — content-testing methods
print("12345".isdigit())    # True  — all characters are digits
print("abc12".isdigit())    # False — contains non-digit characters
print("Python".isalpha())   # True  — all letters
print("  \t\n".isspace())   # True  — all whitespace
print("".isdigit())         # False — empty string returns False

# Practical: validate before converting
raw = input("Enter a number: ")
if raw.strip().isdigit():
    number = int(raw.strip())
    print("Doubled:", number * 2)
else:
    print("That is not a valid integer.")
````

## 2.9 Counting and Checking: `count()`

`count(substring)` returns how many times a substring appears in the string.

````python
# Example 22 — count()
text = "to be or not to be, that is the question"

print(text.count("be"))    # 2
print(text.count("the"))   # 2  ("the" and inside "that"? No — exact match only)
print(text.count("to"))    # 2

# Count vowels in a word
word = "Artificial Intelligence"
vowel_count = (word.lower().count("a") + word.lower().count("e") +
               word.lower().count("i") + word.lower().count("o") +
               word.lower().count("u"))
print(f"Vowels in '{word}': {vowel_count}")
````

## 2.10 Chaining Methods

Because each string method returns a new string, you can **chain** multiple methods in one expression.  
Read the chain left to right — each method acts on the result of the previous one.

````python
# Example 23 — method chaining for input cleaning
raw = "  AHMED AL-MANSOURI  "

# Step by step:
step1 = raw.strip()          # remove surrounding spaces
step2 = step1.lower()        # normalise to lowercase
step3 = step2.replace("-", " ")  # replace hyphens with spaces
step4 = step3.title()        # capitalise each word
print(step4)                 # Ahmed Al Mansouri

# All in one chain:
clean_name = raw.strip().lower().replace("-", " ").title()
print(clean_name)            # Ahmed Al Mansouri
````

# Part 3 — String Formatting in Depth: `.format()`

## 3.1 Why Use `.format()`?

Python has three main ways to embed variables into strings:

| Approach | Syntax | When to use |
|---|---|---|
| Concatenation | `"Hi " + name` | Simple, but cluttered with many variables |
| f-string | `f"Hi {name}"` | Most concise; preferred in modern Python |
| `.format()` | `"Hi {}".format(name)` | Common in older code; powerful for reusable templates |

`.format()` is worth knowing because you will encounter it in documentation, libraries, and code written before Python 3.6.

## 3.2 Basic Placeholders `{}`

Each `{}` in the template is a placeholder. `.format()` fills them in **left to right**.

````python
# Example 24 — basic positional placeholders
name  = "Sara"
score = 94

msg = "{} scored {}% on the exam.".format(name, score)
print(msg)   # Sara scored 94% on the exam.

# Three placeholders — filled in order
report = "Student: {} | Course: {} | Grade: {}".format("Ali", "AI", "A")
print(report)
````

## 3.3 Positional Indexes `{0}`, `{1}`, …

You can refer to arguments by their **position number** inside the braces.  
This lets you reuse or reorder arguments.

````python
# Example 25 — positional indexes
template = "{0} and {1} are studying {2}. {1} also studies {3}."
result = template.format("Aisha", "Omar", "Python", "Data Science")
print(result)
# Aisha and Omar are studying Python. Omar also studies Data Science.

# Reuse the same argument multiple times
announcement = "Attention {0}! {0}, please report to the front desk."
print(announcement.format("Layla"))
````

## 3.4 Named Placeholders `{name}`

Named placeholders make templates self-documenting — the template reads like natural language.

````python
# Example 26 — named placeholders
template = "Dear {student}, your score in {course} is {score}%."

msg1 = template.format(student="Aisha", course="AI",     score=96)
msg2 = template.format(student="Omar",  course="Python", score=88)

print(msg1)
print(msg2)
````

````python
# Example 27 — reusable email template
email_template = (
    "Dear {name},\n\n"
    "Your application for {program} has been {status}.\n"
    "Please contact {office} for next steps.\n\n"
    "Best regards,\nAdmissions Office"
)

email1 = email_template.format(
    name="Aisha Al-Rashid",
    program="MSc Artificial Intelligence",
    status="accepted",
    office="admissions@university.edu"
)

print(email1)
````

## 3.5 Number Formatting

Inside `{}` you can add a **format specification** after a colon: `{value:spec}`.  

| Spec | Meaning | Example |
|---|---|---|
| `.2f` | Float with 2 decimal places | `19.99` |
| `.4f` | Float with 4 decimal places | `3.1416` |
| `d` | Integer | `42` |
| `,` | Thousands separator | `1,000,000` |
| `,.2f` | Thousands + 2 decimals | `1,234.57` |
| `.2%` | Percentage (multiplies by 100) | `94.50%` |
| `e` | Scientific notation | `1.5e+03` |

````python
# Example 28 — number formatting specs
price    = 19.987
pi       = 3.14159265
salary   = 125000
accuracy = 0.9453

print("Price:    {:.2f}".format(price))       # 19.99
print("Pi:       {:.4f}".format(pi))          # 3.1416
print("Salary:   {:,}".format(salary))        # 125,000
print("Budget:   {:,.2f}".format(salary * 3)) # 375,000.00
print("Accuracy: {:.2%}".format(accuracy))    # 94.53%
print("Scientific: {:.2e}".format(0.000123))  # 1.23e-04
````

## 3.6 Width and Alignment

You can control the **minimum width** and **alignment** of a field, which is useful for building tables.

| Alignment char | Meaning |
|---|---|
| `<` | Left-align (default for strings) |
| `>` | Right-align (default for numbers) |
| `^` | Centre-align |

Syntax: `{value:<width}` or `{value:>width}` or `{value:^width}`

````python
# Example 29 — width and alignment
print("{:<20}".format("left"))     # 'left                '
print("{:>20}".format("right"))    # '               right'
print("{:^20}".format("centre"))   # '       centre       '

# Fill character before the alignment symbol
print("{:*^20}".format("Python"))  # *******Python*******
print("{:->20}".format("END"))     # -----------------END
````

````python
# Example 30 — building a simple report table using .format()
header = "{:<15} {:>8} {:>8}".format("Name", "Score", "Grade")
divider = "-" * len(header)

row1 = "{:<15} {:>8.1f} {:>8}".format("Aisha",  96.5, "A")
row2 = "{:<15} {:>8.1f} {:>8}".format("Omar",   88.0, "B")
row3 = "{:<15} {:>8.1f} {:>8}".format("Layla",  74.3, "C")

print(header)
print(divider)
print(row1)
print(row2)
print(row3)
````

## 3.7 `format_map()` and Template Reuse

`format_map()` is a variant that accepts a **dictionary** of values instead of keyword arguments.  
It is useful when you already have data stored in a dictionary.

````python
# Example 31 — format_map() with a dictionary
student = {"name": "Noura", "course": "Python", "score": 91}

msg = "Student {name} completed {course} with a score of {score}.".format_map(student)
print(msg)
# Student Noura completed Python with a score of 91.
````

# Part 4 — Unicode Awareness

## 4.1 What Is Unicode?

**Unicode** is an international standard that assigns a unique number — called a **code point** — to every character in every writing system on Earth: Latin, Arabic, Chinese, emoji, mathematical symbols, and more.

Python 3 strings are **Unicode by default**. You can include any Unicode character directly:

````python
# Example 32 — Python strings are Unicode
english  = "Hello, world!"
arabic   = "\u0645\u0631\u062d\u0628\u0627\u060c \u0639\u0627\u0644\u0645"
emoji    = "Python is fun \U0001F40D"
mixed    = "Caf\u00e9 Python \u2615"

print(english)
print(arabic)    # مرحبا، عالم
print(emoji)     # Python is fun 🐍
print(mixed)     # Café Python ☕

# len() counts code points, not bytes
print(len(arabic))   # number of Arabic characters
````

````python
# Example 33 — ord() and chr(): code points
# ord() converts a character to its Unicode code point number
print(ord("A"))        # 65
print(ord("a"))        # 97
print(ord("م"))        # 1605  (Arabic letter meem)

# chr() converts a code point number back to a character
print(chr(65))         # A
print(chr(97))         # a
print(chr(1605))       # م

# The Latin alphabet occupies code points 65-90 (upper) and 97-122 (lower)
print(chr(65), chr(66), chr(67))   # A B C
````

## 4.2 Encoding and Decoding: `encode()` and `decode()`

A string in memory is a sequence of Unicode code points. When you save text to a file or send it over a network, it must be converted to **bytes** — a process called **encoding**.

Reading bytes back into a string is called **decoding**.

```
string  ──encode()──▶  bytes
bytes   ──decode()──▶  string
```

**UTF-8** is the dominant encoding: it represents every Unicode character using 1–4 bytes and is fully backward-compatible with ASCII.

````python
# Example 34 — encode() and decode()
text = "Café Python"

# Encode the string to bytes using UTF-8
encoded = text.encode("utf-8")
print(type(encoded))    # <class 'bytes'>
print(encoded)          # b'Caf\xc3\xa9 Python'  (é becomes two bytes in UTF-8)
print(len(text))        # 11 characters
print(len(encoded))     # 12 bytes (é takes 2 bytes)

# Decode the bytes back to a string
decoded = encoded.decode("utf-8")
print(decoded)          # Café Python
print(decoded == text)  # True
````

````python
# Example 35 — different encodings produce different bytes
text = "Python"

utf8    = text.encode("utf-8")
utf16   = text.encode("utf-16")
ascii_  = text.encode("ascii")

print("UTF-8:  ", utf8,   len(utf8),   "bytes")
print("UTF-16: ", utf16,  len(utf16),  "bytes")
print("ASCII:  ", ascii_, len(ascii_), "bytes")
````

## 4.3 Encoding Errors

If you try to encode a character that the chosen encoding does not support, Python raises a `UnicodeEncodeError`.  
This is common when using `ascii` or `latin-1` with non-Latin characters.

````python
# Example 36 — encoding errors and how to handle them
arabic_word = "مرحبا"

# UTF-8 handles Arabic fine
utf8_bytes = arabic_word.encode("utf-8")
print("UTF-8 bytes:", utf8_bytes)

# ASCII cannot represent Arabic — this would crash:
# arabic_word.encode("ascii")   # UnicodeEncodeError!

# The errors= parameter controls how to handle failures:
# 'replace' substitutes ? for unencodable characters
safe = arabic_word.encode("ascii", errors="replace")
print("ASCII with replace:", safe)   # b'?????'

# 'ignore' silently drops unencodable characters
ignored = arabic_word.encode("ascii", errors="ignore")
print("ASCII with ignore:", ignored)   # b''
````

## 4.4 Unicode Normalisation

The same visible character can sometimes be represented as **different sequences of code points**.  
For example, `é` can be a single code point **or** the letter `e` followed by a combining accent.

The `unicodedata` module provides `normalize()` to make comparisons reliable.

````python
# Example 40 — why two strings that look the same may not be equal
import unicodedata

# é as a single precomposed character (U+00E9)
e_composed   = "\u00e9"        # é

# é as e + combining accent (U+0065 + U+0301)
e_decomposed = "\u0065\u0301"  # e + ́

print(e_composed)             # é
print(e_decomposed)           # é  — looks the same on screen!
print(len(e_composed))        # 1  — one code point
print(len(e_decomposed))      # 2  — two code points
print(e_composed == e_decomposed)   # False  — different bytes!

# Normalise both to NFC (composed form) before comparing
nfc_a = unicodedata.normalize("NFC", e_composed)
nfc_b = unicodedata.normalize("NFC", e_decomposed)
print(nfc_a == nfc_b)         # True  — safe to compare now
````
