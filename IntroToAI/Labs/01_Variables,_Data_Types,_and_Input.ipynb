## 1. Variables

In Python, a variable is a named location used to store data in the memory. It's like a container that holds a value. Variables allow you to store and manipulate data efficiently.

### Assignment

Variables are created when you assign a value to them using the `=` operator.

```python
# Assigning different types of values to variables
my_string = "Hello, Python!"
my_integer = 100
my_float = 3.14159

print("String variable:", my_string)
print("Integer variable:", my_integer)
print("Float variable:", my_float)
```

### Variables can change

One of the key characteristics of variables is that their values can be changed (re-assigned) during the program's execution. Python is dynamically typed, meaning you can even change the type of value a variable holds.

```python
my_variable = "Initial value"
print("Before change:", my_variable)

my_variable = 25
print("After changing to an integer:", my_variable)
```

### Naming Variables: Rules and Conventions

When naming variables, follow these rules and conventions:

*   **Rules (must follow):**
    *   Names can contain letters (a-z, A-Z), numbers (0-9), and underscores (`_`).
    *   Names cannot start with a number.
    *   Names are case-sensitive (`myVar` is different from `myvar`).
    *   Reserved keywords (like `if`, `else`, `while`, `for`, `class`, `def`, etc.) cannot be used as variable names.

*   **Conventions (good practices):**
    *   Use descriptive names (e.g., `total_sales` instead of `ts`).
    *   Use `snake_case` for variable names (all lowercase, words separated by underscores).
    *   For constants (values that shouldn't change), use `SCREAMING_SNAKE_CASE` (all uppercase with underscores).


```python
# Good variable names
user_name = "Alice"
product_price = 19.99
is_active = True

# Example of a constant
MAX_RETRIES = 5

print("User:", user_name, ", Price:", product_price, ", Active:", is_active, ", Max Retries:", MAX_RETRIES)
```

## 2. Data Types

Python has several built-in data types to categorize values. Here are some of the most common ones:

### Integer (`int`)

Represents whole numbers, positive or negative, without a decimal point.

```python
age = 30
quantity = -5
big_number = 1000000000
print("Age:", age, ", Quantity:", quantity, ", Big Number:", big_number)
```

### Float (`float`)

Represents real numbers, with a decimal point or in exponential form.

```python
price = 29.99
temperature = -10.5
scientific_notation = 1.23e-5 # 0.0000123
print("Price:", price, ", Temperature:", temperature, ", Scientific:", scientific_notation)
```

### Boolean (`bool`)

Represents truth values: `True` or `False`.

```python
is_admin = True
has_permission = False
print("Is Admin:", is_admin, ", Has Permission:", has_permission)
```

### String (`str`)

Represents sequences of characters, enclosed in single (`'`) or double (`"`) quotes.

```python
name = 'Charlie'
message = "Welcome to Python!"
multiline_string = """This is a
multi-line string."""
print("Name:", name, ", Message:", message + "\n" + multiline_string)
```

### NoneType (`None`)

Represents the absence of a value. It's often used as a placeholder for a value that doesn't exist yet.

```python
result = None
user_data = None # Will be populated later
print("Result:", result, ", User Data:", user_data)
```

## 3. Checking Data Types

### `type()` function

The `type()` function returns the type of an object.

```python
num = 10
text = "Python"
decimal = 5.5
bool_val = True
nil_val = None

print("Type of num:", type(num))
print("Type of text:", type(text))
print("Type of decimal:", type(decimal))
print("Type of bool_val:", type(bool_val))
print("Type of nil_val:", type(nil_val))
```

### `isinstance()` function

The `isinstance()` function checks if an object is an instance of a specified class (or a tuple of classes). It's generally preferred over `type()` for type checking, especially with inheritance.

```python
value = "Hello"

print("Is '", value, "' a string?", isinstance(value, str))
print("Is '", value, "' an integer?", isinstance(value, int))
print("Is 10 an integer?", isinstance(10, int))
print("Is 10.0 a float?", isinstance(10.0, float))
```

## 4. Type Conversion (Casting)

Type conversion, or casting, is the process of converting a value from one data type to another. Python provides built-in functions for this (`int()`, `float()`, `str()`, `bool()`, etc.).

### Examples of successful casting

```python
str_num = "123"
int_num = int(str_num)
print("String '", str_num, "' to int:", int_num, ", type:", type(int_num))

float_str = "45.67"
float_num = float(float_str)
print("String '", float_str, "' to float:", float_num, ", type:", type(float_num))

int_to_float = float(10)
print("Int 10 to float:", int_to_float, ", type:", type(int_to_float))

float_to_int = int(9.81) # Truncates the decimal part
print("Float 9.81 to int:", float_to_int, ", type:", type(float_to_int))

int_to_str = str(500)
print("Int 500 to string: '", int_to_str, "', type:", type(int_to_str))

# Boolean conversion
print("Bool of 0:", bool(0)) # False
print("Bool of 1:", bool(1)) # True
print("Bool of empty string:", bool('')) # False
print("Bool of non-empty string:", bool('abc')) # True
```

### When casting fails

Casting will fail if the value cannot be logically converted to the target type. For example, trying to convert a non-numeric string to an integer or float will raise a `ValueError`.

```python
# The following lines will cause a ValueError if uncommented and run,
# because 'hello' cannot be converted to an integer and '12.3.4' is not a valid float.
# invalid_int = int("hello")
# invalid_float = float("12.3.4")
print("Attempting to convert 'hello' to int or '12.3.4' to float will result in a ValueError.")
```

## 5. The `len()` function

The `len()` function returns the number of items in an object. It works for sequences (like strings, lists, tuples) and collections (like dictionaries, sets).

```python
my_string = "Python Programming"

print("Length of string '", my_string, "':", len(my_string))
```

## 6. Printing Variables: Formatted F-strings

F-strings (formatted string literals) provide a concise and convenient way to embed Python expressions inside string literals for formatting. They are introduced with an `f` or `F` before the opening quotation mark.

```python
name = "Bob"
score = 95.5
rank = 3

# Using f-strings to combine text and variables
print(f"Player: {name}, Score: {score}, Rank: {rank}")

# F-strings can include expressions
print(f"Next rank would be: {rank + 1}")

# Formatting floats
pi = 3.1415926535
print(f"Pi with 2 decimal places: {pi:.2f}")
print(f"Pi with 4 decimal places: {pi:.4f}")
```

## 7. The `input()` function

The `input()` function allows your program to take input from the user via the keyboard. It reads a line from input, converts it to a string, and returns that string.

### Basic usage

```python
# Prompt the user for their name
user_name = input("Enter your name: ")
print(f"Hello, {user_name}!")

# The input is always a string
print(f"Type of user_name: {type(user_name)}")
```

### Using `input()` with Type Conversions

Since `input()` always returns a string, you often need to convert the input to the desired data type (e.g., `int` or `float`) if you plan to perform numerical operations.

**Warning**: Please ensure you enter valid numbers for the prompts below. Entering non-numeric text will cause a `ValueError` and stop the execution of this cell.

Since `input()` always returns a string, you often need to convert the input to the desired data type (e.g., `int` or `float`) if you plan to perform numerical operations.

```python
# Get age and convert to integer
age_str = input("Enter your age: ")
age = int(age_str)
print(f"You are {age} years old.")
print(f"In 5 years, you will be {age + 5} years old.")
print(f"Type of age: {type(age)}")

# Get a decimal number and convert to float
price_str = input("Enter a price: ")
price = float(price_str)
print(f"The price is ${price:.2f}")
print(f"Type of price: {type(price)}")
```
