## 1. What Is a Loop?

A **loop** is a programming construct that repeats a block of code multiple times. Instead of writing the same code over and over, you write it once and let the loop do the repetition.

Python has two kinds of loops:

| Loop | Best used when… |
|------|-----------------|
| `for` | You know what you want to iterate *over* (a sequence, a range, etc.) |
| `while` | You want to keep going *while* some condition is true |

---

## 2. The `for` Loop

### 2.1 Iterating Over a String

A string is a sequence of characters. The `for` loop visits each character one at a time.

````python
# Print each letter in a word on its own line
word = "Python"
for letter in word:
    print(letter)
````

**How it works:**
- Before each iteration, the *next* character in `word` is assigned to `letter`
- The loop body runs once per character
- When there are no more characters, the loop ends

````python
# Print only the consonants in a word (skip vowels)
word = "Programming"
for char in word:
    if char not in "aeiouAEIOU":
        print(char, end=" ")
````

### 2.2 Iterating Over a Range

The `range()` function generates a sequence of integers (covered in detail in Section 4).

````python
# Count from 1 to 5
for i in range(1, 6):
    print(i)
````

````python
# Sum the numbers 1 through 10
total = 0
for i in range(1, 11):
    total = total + i

print("Sum from 1 to 10:", total)
````

### 2.3 Using `len()` to Iterate by Index

Sometimes you need the *position* of each character, not just the character itself. Combine `range()` with `len()`:

````python
name = "KFUPM"
for i in range(len(name)):
    print("Position", i, "→", name[i])
````

````python
# Print the position of each uppercase letter in a sentence
sentence = "Hello World from Python"
for i in range(len(sentence)):
    if sentence[i].isupper():
        print("Uppercase letter", sentence[i], "at position", i)
````

## 3. The `while` Loop

```
while condition:
    body
```

The body keeps running **as long as** the condition is `True`. When the condition becomes `False`, the loop ends.

> ⚠️ **Infinite loop warning:** Always make sure something inside the loop eventually makes the condition `False`, otherwise the program will run forever.

### 3.1 Count-Controlled `while` Loop

````python
# Print 1 through 5 using a while loop
counter = 1
while counter <= 5:
    print(counter)
    counter = counter + 1   # don't forget to update!
````

Every count-controlled loop has three parts:

| Part | Example above |
|------|---------------|
| **Initialize** | `counter = 1` |
| **Check** | `counter <= 5` |
| **Update** | `counter = counter + 1` |

````python
# Keep doubling a number until it exceeds 100
n = 1
while n < 100:
    n = 2 * n
    print(n)
````

### 3.2 Event-Controlled `while` Loop

Sometimes the loop runs until an *event* occurs — like reaching a financial goal.

````python
# How many years to double a bank balance at 5% annual interest?
RATE = 5
balance = 10_000.0
TARGET = 2 * balance
year = 0

while balance < TARGET:
    year = year + 1
    interest = balance * RATE / 100
    balance = balance + interest

print(f"The investment doubled after {year} years.")
print(f"Final balance: ${balance:.2f}")
````

### 3.3 Sentinel-Controlled `while` Loop

A **sentinel value** is a special value that signals "I'm done entering data".

````python
# Keep reading grades until the user enters -1
total = 0.0
count = 0
grade = float(input("Enter a grade (or -1 to stop): "))

while grade != -1:
    total = total + grade
    count = count + 1
    grade = float(input("Enter a grade (or -1 to stop): "))

if count > 0:
    print("Average grade:", total / count)
else:
    print("No grades were entered.")
````

````python
# Running total — loop stops once total reaches 10
i = 0
total = 0
while total < 10:
    i = i + 1
    total = total + i
    print("i =", i, "  total =", total)
````

## 4. The `range()` Function

`range()` produces a sequence of integers. It takes up to three arguments:

```python
range(stop)              # 0, 1, 2, ..., stop-1
range(start, stop)       # start, start+1, ..., stop-1
range(start, stop, step) # start, start+step, ..., up to (not including) stop
```

> 📌 The **stop** value is **never** included.

### 4.1 One Argument — `range(stop)`

````python
for i in range(5):
    print(i, end=" ")   # prints 0 1 2 3 4
````

### 4.2 Two Arguments — `range(start, stop)`

````python
for i in range(3, 8):
    print(i, end=" ")   # prints 3 4 5 6 7
````

### 4.3 Three Arguments — `range(start, stop, step)`

````python
# Even numbers from 2 to 10
print("Even numbers:")
for i in range(2, 11, 2):
    print(i, end=" ")

print()  # new line

# Odd numbers from 1 to 9
print("Odd numbers:")
for i in range(1, 10, 2):
    print(i, end=" ")
````

````python
# Countdown using a negative step
print("Countdown:")
for i in range(10, 0, -1):
    print(i, end=" ")
print("Liftoff! 🚀")
````

````python
# Multiples of 3 from 3 to 30
for i in range(3, 31, 3):
    print(i, end=" ")
````

### 4.4 Quick Reference Table

| Expression | Values produced |
|------------|------------------|
| `range(5)` | 0, 1, 2, 3, 4 |
| `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(0, 10, 2)` | 0, 2, 4, 6, 8 |
| `range(10, 0, -1)` | 10, 9, 8, …, 1 |
| `range(5, 5)` | *(empty — start equals stop)* |

## 5. Loop Control: `break`, `continue`, and `pass`

These three keywords give you fine-grained control over what happens inside a loop.

---

### 5.1 `break` — Exit the Loop Immediately

`break` stops the loop and jumps to the first statement *after* the loop body.

````python
# Find the first digit in a string
text = "Hello3World"

for char in text:
    if char.isdigit():
        print("First digit found:", char)
        break   # no need to keep going

print("Loop finished.")
````

````python
# Keep asking for a password until the user gets it right
secret = "python123"

while True:   # loop forever... until a break
    attempt = input("Enter the password: ")
    if attempt == secret:
        print("Access granted! ✅")
        break
    else:
        print("Wrong password. Try again.")
````

---

### 5.2 `continue` — Skip to the Next Iteration

`continue` skips the rest of the current iteration and jumps back to the loop condition (or next item).

````python
# Print numbers 1-10, but skip multiples of 3
for i in range(1, 11):
    if i % 3 == 0:
        continue   # skip this iteration
    print(i, end=" ")
````

````python
# Count only the uppercase letters in a sentence
sentence = "Hello World from KFUPM"
count = 0

for char in sentence:
    if not char.isupper():
        continue          # skip non-uppercase characters
    count = count + 1

print("Uppercase letters:", count)
````

---

### 5.3 `pass` — Do Nothing (Placeholder)

`pass` is a no-op — it does absolutely nothing. It is used when Python requires a statement syntactically but you don't want any action.

````python
# Placeholder loop body while you plan the logic later
for i in range(5):
    pass   # TODO: fill this in later

print("Loop ran without errors.")
````

````python
# Extract only the digits from a string, ignore everything else
text = "ICS 104"
digits_found = ""

for char in text:
    if char.isalpha() or char == " ":
        pass   # ignore letters and spaces
    else:
        digits_found = digits_found + char

print("Digits extracted:", digits_found)
````

### Quick Comparison

| Keyword | Effect |
|---------|--------|
| `break` | Exits the entire loop immediately |
| `continue` | Skips the rest of *this* iteration; moves to the next |
| `pass` | Does nothing; used as a placeholder |

## 6. The `else` Clause on Loops

This is one of Python's most **overlooked** features — and one of its most useful.

Both `for` and `while` loops can have an `else` block:

```python
for item in sequence:
    ...
else:
    # runs if the loop completed WITHOUT hitting a break
    ...
```

**Rule:** The `else` block runs when the loop *finishes normally* (exhausts all items or condition becomes `False`). It does **not** run if the loop ended via `break`.

### 6.1 Searching for Something

````python
# Search for a digit — no digit present, so else runs
text = "HelloWorld"

for char in text:
    if char.isdigit():
        print("Digit found:", char)
        break
else:
    # Only reaches here if break was never triggered
    print("No digit found in the string.")
````

````python
# Same search — digit IS present, so else does NOT run
text = "Hello3World"

for char in text:
    if char.isdigit():
        print("Digit found:", char)
        break
else:
    print("No digit found in the string.")
````

````python
# Check if a string contains any uppercase letter
text = "hello world"

for char in text:
    if char.isupper():
        print("Uppercase letter found:", char)
        break
else:
    print("No uppercase letters found.")
````

### 6.2 `else` on a `while` Loop

````python
# Validate a PIN: user gets 3 attempts
PIN = "1234"
attempts = 3

while attempts > 0:
    guess = input(f"Enter PIN ({attempts} attempts left): ")
    if guess == PIN:
        print("Correct PIN! Access granted. ✅")
        break
    attempts = attempts - 1
else:
    # Loop exhausted all attempts without a break
    print("Too many wrong attempts. Account locked. 🔒")
````

## 7. Nested Loops

A **nested loop** is a loop inside another loop. The inner loop completes all of its iterations for *each single iteration* of the outer loop.

### 7.1 A Simple Example

````python
# Print a 3x4 grid of coordinates
for row in range(1, 4):          # outer loop: 3 rows
    for col in range(1, 5):      # inner loop: 4 columns
        print(f"({row},{col})", end="  ")
    print()   # move to the next line after each row
````

How many times does the inner loop body run?  
**Outer iterations × Inner iterations = 3 × 4 = 12 times**

### 7.2 Multiplication Table

````python
# Print a 5x5 multiplication table
print("Multiplication Table (1–5)")
print("-" * 30)

for i in range(1, 6):            # row
    for j in range(1, 6):        # column
        print(f"{i*j:4d}", end="")
    print()   # new line after each row
````

### 7.3 Pattern Printing

````python
# Print a right triangle of stars
rows = 5
for i in range(1, rows + 1):     # outer: each row
    for j in range(i):           # inner: print i stars
        print("*", end="")
    print()                      # new line
````

````python
# Print a number triangle
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
````

### 7.4 ⚠️ Performance Awareness

Nested loops multiply their iteration counts. Be careful with large ranges:

````python
# How many times does the inner body execute?
count = 0
for i in range(100):        # 100 iterations
    for j in range(100):    # 100 iterations each
        count = count + 1

print("Inner body ran", count, "times")   # 10,000!
````

With three levels of nesting and `range(100)` each, the body would run **1,000,000** times. Always ask: *do I really need this many iterations?*

## 8. Putting It All Together

### 8.1 Prime Number Checker

This combines a `for` loop, `break`, and the `else` clause elegantly.

````python
# Check whether a number is prime
number = int(input("Enter a positive integer: "))

for divisor in range(2, number):
    if number % divisor == 0:
        print(number, "is NOT prime (divisible by", str(divisor) + ")")
        break
else:
    # Loop finished without finding a divisor
    if number > 1:
        print(number, "IS prime ✅")
    else:
        print(number, "is not considered prime")
````

### 8.2 Character Frequency Counter

Count how many times each unique character appears in a string using only loops and strings.

````python
text = "mississippi"
checked = ""   # keep track of characters we've already counted

for char in text:
    if char in checked:
        continue   # already counted this one

    # Count occurrences of char
    count = 0
    for c in text:
        if c == char:
            count = count + 1

    print(f"'{char}' appears {count} time(s)")
    checked = checked + char   # mark as counted
````

### 8.3 Phone Number Validator

Validate a Saudi mobile number in the format `05XXXXXXXX` using a loop and `break`/`else`.

````python
phone = "0512345678"

# Rules: exactly 10 chars, starts with '05', rest are digits
valid = len(phone) == 10 and phone[0] == "0" and phone[1] == "5"

if valid:
    for i in range(2, len(phone)):
        if not phone[i].isdigit():
            valid = False
            break

if valid:
    print(phone, "is a valid Saudi mobile number ✅")
else:
    print(phone, "is NOT a valid Saudi mobile number ❌")
````

### 8.4 FizzBuzz

````python
# FizzBuzz: for multiples of 3 print "Fizz", multiples of 5 print "Buzz",
# multiples of both print "FizzBuzz", otherwise print the number
for n in range(1, 31):
    if n % 3 == 0 and n % 5 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)
````

## 9. Common Mistakes to Avoid

### ❌ Off-by-One Errors

Think carefully about whether your range should use `<` or `<=`, and whether it starts at 0 or 1.

````python
# Wrong: prints 0 through 4 (not 5 through 10)
print("Wrong:", end=" ")
for i in range(5):
    print(i, end=" ")

print()

# Correct: prints 5 through 10
print("Correct:", end=" ")
for i in range(5, 11):
    print(i, end=" ")
````

### ❌ Forgetting to Update the Loop Variable (Infinite Loop)

````python
# ⛔ DO NOT RUN THIS — it loops forever!
# i = 0
# while i < 5:
#     print(i)         # i never changes → infinite loop!

# ✅ Correct version:
i = 0
while i < 5:
    print(i)
    i = i + 1   # update the variable
````

### ❌ Modifying the Loop Variable Inside a `for` Loop

````python
# This does NOT skip numbers — Python reassigns i each iteration anyway
for i in range(1, 6):
    i = i * 10   # this change is discarded at the start of the next iteration
    print(i)

# Output: 10 20 30 40 50 — i is still 1,2,3,4,5 from range(); *10 happens fresh each time
````
