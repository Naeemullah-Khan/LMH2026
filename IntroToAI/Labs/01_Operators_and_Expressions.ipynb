## 0 · What Is an Operator?

An **operator** is a compact symbol (or keyword) that performs an operation on one or more values, called **operands**. You can think of it as a built-in function wrapped in a concise syntax.

For example, the addition operator `+` could be written as a function call:

```
add(2, 3)  →  5          # imaginary function
2 + 3      →  5          # real operator syntax
```

Both achieve the same thing. Operators just make the code shorter and more readable.

### Arity — how many operands does an operator need?

| Arity | Name | Example |
|-------|------|---------|
| 1 | Unary | `-x`, `not x` |
| 2 | Binary | `x + y`, `x == y` |

### The five operator families we'll cover

1. **Arithmetic** — math operations (`+`, `-`, `*`, `/`, …)
2. **Comparison** — compare values, return `True`/`False` (`==`, `>`, `<=`, …)
3. **Assignment** — update a variable in place (`+=`, `-=`, …)
4. **Other Boolean** — membership & identity (`in`, `is`, `not`)
5. **Logical** — combine boolean results (`and`, `or`)

## 1 · Arithmetic Operators

The building blocks of numeric computation.

### 1.1 Binary arithmetic

````python
print(5 + 2)   # addition       → 7
print(5 - 2)   # subtraction    → 3
print(5 * 2)   # multiplication → 10
print(5 / 2)   # division       → 2.5  (always float in Python 3)
print(5 ** 2)  # exponentiation → 25  (5²)
````

### 1.2 Floor division and modulus

Two operators that you might not have met in everyday maths but are essential in programming.

````python
# Floor division (//) — divide then round DOWN to the nearest whole number
print(5 // 2)   # → 2  (not 2.5)
print(7 // 3)   # → 2
print(9 // 3)   # → 3
````

````python
# Modulus (%) — the *remainder* after division
print(4 % 2)    # → 0  (4 divides evenly)
print(5 % 2)    # → 1
print(10 % 3)   # → 1  (10 = 3*3 + 1)
print(11 % 4)   # → 3  (11 = 4*2 + 3)
````

💡 **Tip — Even/odd check with `%`:**
```python
number % 2 == 0   # True  → even
number % 2 == 1   # True  → odd
```

### 1.3 Unary arithmetic

````python
x = 5
print(-x)   # negation  → -5
print(x)    # x itself is unchanged → 5
````

### 1.4 Operators work on other types too (operator overloading)

Python lets classes define how operators behave — this is called **operator overloading**. You'll explore it fully in Advanced OOP, but you already benefit from it every day:

````python
# Strings — + concatenates, * repeats
print("Hello" + " " + "World")   # → Hello World
print("ha" * 3)                  # → hahaha

# Lists — + joins, * repeats
print([1, 2] + [3, 4])           # → [1, 2, 3, 4]
print([0] * 4)                   # → [0, 0, 0, 0]
````

## 2 · Comparison Operators

Comparison operators compare two values and **always return a boolean** (`True` or `False`).

### 2.1 Equality and inequality (`==`, `!=`)

````python
# Numbers
print(4 == 4)   # True  — equal
print(4 == 3)   # False
print(4 != 3)   # True  — not equal
print(4 != 4)   # False
````

````python
# Strings — comparison is case-sensitive!
print('a' == 'a')   # True
print('a' == 'A')   # False  ← different ASCII values
print('a' != 'A')   # True
````

### 2.2 Greater than / greater than or equal to (`>`, `>=`)

````python
print(4 > 3)    # True
print(4 > 4)    # False — strictly greater
print(4 > 5)    # False

print(4 >= 3)   # True
print(4 >= 4)   # True  — equal counts!
print(4 >= 5)   # False
````

### 2.3 Less than / less than or equal to (`<`, `<=`)

````python
print(4 < 5)    # True
print(4 < 4)    # False
print(4 < 3)    # False

print(4 <= 5)   # True
print(4 <= 4)   # True
print(4 <= 3)   # False
````

### 2.4 Chaining comparisons

Python lets you chain comparisons naturally — just like in maths:

````python
age = 25

# Is age between 18 and 65?
print(18 <= age <= 65)   # True — much cleaner than writing age >= 18 and age <= 65
````

## 3 · Assignment Operators

Assignment operators are shorthand for **performing an operation and saving the result back** into the same variable.

```python
x = x + 1   # verbose form
x += 1       # shorthand — identical result
```

Every arithmetic operator has a matching assignment operator:

````python
x = 10
x += 3    # x = x + 3
print("+=  →", x)   # 13
````

````python
x = 10
x -= 3    # x = x - 3
print("-=  →", x)   # 7
````

````python
x = 10
x *= 3    # x = x * 3
print("*=  →", x)   # 30
````

````python
x = 10
x /= 4    # x = x / 4
print("/=  →", x)   # 2.5
````

````python
x = 10
x //= 3   # x = x // 3
print("//= →", x)   # 3
````

````python
x = 3
x **= 3   # x = x ** 3  (3 cubed)
print("**= →", x)   # 27
````

````python
x = 17
x %= 5    # x = x % 5
print("%=  →", x)   # 2
````

💡 **Real-world use case:** counting iterations, accumulating totals, or tracking a running balance are all perfect jobs for `+=` and `-=`.

```python
score = 0
score += 10   # player scores a point
score += 5    # bonus
score -= 3    # penalty
print(score)  # 12
```

## 4 · Other Boolean Operators

These operators are used for membership checks, identity checks, and negation.

### 4.1 Membership operator — `in`

Checks whether a value exists inside a collection (string, list, tuple, set, dict key, …).

````python
# In a string
print('e' in 'Hello World')    # True
print('x' in 'Hello World')    # False
print('World' in 'Hello World') # True — substring check

# In a list
languages = ['Python', 'Ruby', 'Go']
print('Python' in languages)   # True
print('Java' in languages)     # False

# Negated membership
print('Java' not in languages) # True
````

### 4.2 Identity operator — `is`

`is` checks whether two names point to the **exact same object in memory** — not just equal values.

> ⚠️ **Rule of thumb:** only use `is` with `True`, `False`, and `None`. For everything else, use `==`.

````python
# Correct uses of `is`
result = None
print(result is None)    # True  ✅

active = True
print(active is True)    # True  ✅

closed = False
print(closed is False)   # True  ✅
````

````python
# Why NOT to use `is` with arbitrary values
a = 1000
b = 1000
print(a == b)   # True  — values are equal
print(a is b)   # May be False! — they could be different objects in memory
````

The three styles for checking a boolean variable — ranked by readability:

```python
if active == True:   # ❌ verbose and redundant
if active is True:   # ✅ readable, explicit
if active:           # ✅✅ most Pythonic (preferred style)
```

### 4.3 Negation operator — `not`

`not` is a **unary** operator — it flips a boolean value.

````python
print(not True)    # False
print(not False)   # True
````

````python
user_is_active = False

# The expression 'not user_is_active' evaluates to True if user_is_active is False
# This demonstrates a "guard clause" concept where an action is taken if a condition is met.
print(f"Is user not active? {not user_is_active}")
````

`not` also works on other types — it treats any "empty" or "zero" value as `False`:

````python
print(not 0)      # True  — zero is falsy
print(not 1)      # False — any non-zero number is truthy
print(not '')     # True  — empty string is falsy
print(not 'hi')   # False — non-empty string is truthy
print(not [])     # True  — empty list is falsy
print(not [1])    # False
````

## 5 · Logical Operators

Logical operators let you **combine** boolean expressions. They read almost like plain English.

### 5.1 `and` — all conditions must be true

| A | B | A `and` B |
|---|---|----------|
| True | True | **True** |
| True | False | False |
| False | True | False |
| False | False | False |

````python
user_is_active = True
age = 21

# The 'and' operator returns True only if both conditions are True.
print(f"Access condition: (user is active and age is >= 18) = {user_is_active and age >= 18}")
````

````python
user_is_active = True
age = 15

# The 'and' operator returns False if any condition is False.
print(f"Access condition: (user is active and age is >= 18) = {user_is_active and age >= 18}")
````

### 5.2 `or` — at least one condition must be true

| A | B | A `or` B |
|---|---|----------|
| True | True | True |
| True | False | True |
| False | True | True |
| False | False | **False** |

````python
food = "spaghetti"

# The 'or' operator returns True if at least one condition is True.
print(f"Do I like this food? (food is hamburger or food is spaghetti) = {food == "hamburger" or food == "spaghetti"}")
````

### 5.3 Combining `and`, `or`, `not`

You can chain multiple logical operators. Use parentheses `()` to make the intent clear (they also control evaluation order).

````python
is_member = True
has_voucher = False
balance = 50

# Combine 'or' and 'and' to evaluate complex conditions.
# (member OR has a voucher) AND balance is sufficient
print(f"Purchase approval condition: {(is_member or has_voucher) and balance >= 20}")
````

## 6 · Operator Precedence

When multiple operators appear in a single expression, Python follows a strict evaluation order — similar to BODMAS/PEMDAS in maths.

| Priority | Operators |
|----------|-----------|
| Highest | `**` |
| | `+x`, `-x` (unary) |
| | `*`, `/`, `//`, `%` |
| | `+`, `-` |
| | `==`, `!=`, `<`, `>`, `<=`, `>=`, `in`, `is` |
| | `not` |
| | `and` |
| Lowest | `or` |

**When in doubt, use parentheses `()`** — they always override precedence and make your intent explicit.

````python
# Precedence in action
print(2 + 3 * 4)        # 14, not 20 — * before +
print((2 + 3) * 4)      # 20 — parentheses first

print(True or False and False)    # True — `and` before `or`
print((True or False) and False)  # False — parentheses change the result!
````

## 8 · Built-in Function Calls

While operators are concise symbols for operations, Python also provides many built-in **functions** that perform more complex or specific tasks. Like operators, functions take arguments (inputs) and return a result. Here are a few common ones:

### `abs()` — Absolute Value

Returns the absolute value of a number (its distance from zero).

````python
print(abs(-5))   # → 5
print(abs(10))   # → 10
````

### `max()` and `min()` — Maximum and Minimum

`max()` returns the largest item in an iterable or the largest of two or more arguments.
`min()` returns the smallest item in an iterable or the smallest of two or more arguments.

````python
print(max(10, 20, 5))    # → 20
print(min(10, 20, 5))    # → 5
print(max([1, 5, 9]))    # → 9 (works with lists too)
````

### `round()` — Rounding Numbers

Returns a number rounded to `ndigits` precision after the decimal point. If `ndigits` is omitted or `None`, it returns the nearest integer.

````python
print(round(3.14159))     # → 3
print(round(3.7))         # → 4
print(round(3.14159, 2))  # → 3.14
````
