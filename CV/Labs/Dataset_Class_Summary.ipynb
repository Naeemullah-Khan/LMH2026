To make a dataset class, you can follow these steps:

1. Explore the directory (Use the file explorer in the left OR `ls`).
2. Collect the image paths + labels/masks_paths (good news: the exam will be similar to the labs, so copy/paste/adapt from the labs).
3. If data is not split: Create a DataFrame with paths/labels and use train_test_split.
4. For each image:
    * **a.** Read the image (use `Image.open(XXX).convert("RGB")`).
    * **b.** Read the label/mask (for the mask, always use `Image.open(XXX).convert("L")`).
    * **c.** Manipulate the dimensions if needed (images should be `CxHxW`).
    * **d.** Apply the transformations (resize all images to the same size, any augmentations, convert to tensor, normalize). **Note:** use `ToTensor` for images, and `PILToTensor` for masks.
    * **e.** Return the image and the label/mask.

Step (2) could be inside the `__init__` or outside the dataset class. Step (4) should be inside the `__getitem__` function.

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("kritikseth/fruit-and-vegetable-image-recognition")     # Data classified into train, val, test

print("Path to dataset files:", path)

```

```python
import os
import glob
from torch.utils.data import Dataset
from PIL import Image
import os
from torchvision.datasets import ImageFolder
from torchvision import transforms

class StandardDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform

        # 1. Auto-generate class dictionary (or define manually)
        # Sort ensures 0=Apple, 1=Banana consistently
        classes = sorted([d for d in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, d))])
        self.class_labels = {cls_name: i for i, cls_name in enumerate(classes)}

        self.image_paths = []
        self.labels = []

        # 2. Collect Paths
        for class_name, label in self.class_labels.items():
            paths = glob.glob(f"{root_dir}/{class_name}/*.*") # Matches .jpg, .png etc
            self.image_paths.extend(paths)
            self.labels.extend([label] * len(paths))

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # 3. Load & Process
        image = Image.open(self.image_paths[idx]).convert("RGB")
        label = self.labels[idx]

        if self.transform:
            image = self.transform(image)

        return image, label

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

train_dir = os.path.join(path, "train")
val_dir = os.path.join(path, "validation")
test_dir = os.path.join(path, "test")

train_dataset = StandardDataset(root_dir=train_dir, transform=transform)
val_dataset = StandardDataset(root_dir=val_dir, transform=transform)
test_dataset = StandardDataset(root_dir=test_dir, transform=transform)


# OR


from torchvision.datasets import ImageFolder

# Automatically handles everything if folders are named correctly
train_dataset = ImageFolder(root=train_dir, transform=transform)
val_dataset   = ImageFolder(root=val_dir,   transform=transform)
test_dataset  = ImageFolder(root=test_dir,  transform=transform)
```

```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split

# 1. Load CSV
# Note: Check the column names to find where is the image name and its label/mask
csv_path = os.path.join(path, "metadata.csv")
df = pd.read_csv(csv_path)

# 2. Create Label Mapping (String -> Number)
# This finds all unique names and numbers them 0, 1, 2...
classes = sorted(df['Font'].unique()) # Replace 'label' with actual column name
class_to_idx = {cls_name: i for i, cls_name in enumerate(classes)}

# 3. Split DataFrame (80% Train, 20% Test)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['Font'])

print(f"Train samples: {len(train_df)}, Test samples: {len(test_df)}")
```

```python
from torch.utils.data import Dataset
from PIL import Image

class FontDataset(Dataset):
    def __init__(self, df, root_dir, class_map, transform=None):
        self.df = df
        self.root_dir = root_dir
        self.class_map = class_map
        self.transform = transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # 1. Get row info
        # iloc[idx] gives the row at position idx
        row = self.df.iloc[idx]

        # 2. Get Filename & Label
        # Replace 'file_name' and 'label' with ACTUAL headers from CSV
        img_name = row['file_name']
        label_str = row['font']

        # 3. Load Image
        img_path = os.path.join(self.root_dir, img_name)
        image = Image.open(img_path).convert("RGB")

        # 4. Convert Label to Number
        label = self.class_map[label_str]

        if self.transform:
            image = self.transform(image)

        return image, label




from torchvision import transforms
from torch.utils.data import DataLoader

# 1. Define Transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# 2. Define Image Directory
img_dir = os.path.join(path, "files")

# 3. Create Datasets (Pass the split DataFrames)
train_dataset = FontDataset(train_df, img_dir, class_to_idx, transform=transform)
test_dataset  = FontDataset(test_df,  img_dir, class_to_idx, transform=transform)

# 4. DataLoaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=32, shuffle=False)


print(f"Train: {len(train_dataset)}, Test: {len(test_dataset)}")
```

```python
from torch.utils.data import Dataset
from PIL import Image
import torch
import numpy as np

def remap_mask(mask):
    # Remaps values to 0, 1, 2... based on unique values present
    mask = mask.long() # Segmentation masks must be LongTensor
    unique_values = torch.unique(mask)
    remapped_mask = torch.zeros_like(mask)

    for new_val, old_val in enumerate(sorted(unique_values.tolist())):
        remapped_mask[mask == old_val] = new_val

    return remapped_mask

class BinarySegDataset(Dataset):
    def __init__(self, df, transform=None, mask_transform=None):
        self.df = df
        self.transform = transform
        self.mask_transform = mask_transform

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        # 1. Get Paths from DataFrame
        row = self.df.iloc[idx]
        img_path = row['image_path']
        mask_path = row['mask_path']

        # 2. Load Image & Mask
        image = Image.open(img_path).convert("RGB")
        mask = Image.open(mask_path).convert("L") # Grayscale for masks

        # 3. Apply Transforms
        if self.transform:
            image = self.transform(image)
        if self.mask_transform:
            mask = self.mask_transform(mask)

        # 3. Remap Mask (we need this because CrossEntropy requires the labels to be consecutive)
        mask = remap_mask(mask)

        return image, mask


        from torchvision import transforms
from torch.utils.data import DataLoader

# 1. Define Transforms
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()         # ToTensor does two things: Convert to tensor + scaling (divide by 255)
])
transform_mask = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.PILToTensor()       # PILToTensor does one thing: Convert to tensor only (mask should not be scaled!!)
])

# 2. Create Datasets
train_dataset = BinarySegDataset(train_df, transform=transform, mask_transform=transform_mask)
test_dataset  = BinarySegDataset(test_df,  transform=transform, mask_transform=transform_mask)

# 3. DataLoaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=16, shuffle=False)

# 4. Check Batch
imgs, masks = next(iter(train_loader))
print(f"Image: {imgs.shape}, Mask: {masks.shape}")
```

```python
import torch
import numpy as np

def remap_mask(mask):
    # Remaps values to 0, 1, 2... based on unique values present
    mask = mask.long() # Segmentation masks must be LongTensor
    unique_values = torch.unique(mask)
    remapped_mask = torch.zeros_like(mask)

    for new_val, old_val in enumerate(sorted(unique_values.tolist())):
        remapped_mask[mask == old_val] = new_val

    return remapped_mask


import os
import glob
from torch.utils.data import Dataset
from PIL import Image

class MulticlassDataset(Dataset):
    def __init__(self, image_paths, mask_paths, transform=None, mask_transform=None):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.transform = transform
        self.mask_transform = mask_transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # 1. Load Image & Mask
        image = Image.open(self.image_paths[idx]).convert("RGB")
        mask = Image.open(self.mask_paths[idx]).convert("L") # Keep as L (grayscale)

        # 2. Apply Transform
        if self.transform:
            image = self.transform(image)
        if self.mask_transform:
            mask = self.mask_transform(mask)

        # 3. Remap Mask (we need this because CrossEntropy requires the labels to be consecutive)
        mask = remap_mask(mask)

        return image, mask



from sklearn.model_selection import train_test_split
from torchvision import transforms
from torch.utils.data import DataLoader

# Let's assume the data is not splitted here, we will use only the training folder, then split it.
# 1. Define Paths
root_dir = os.path.join(path, "train")
all_images = sorted(glob.glob(f"{root_dir}/images/*.jpg"))
all_masks  = sorted(glob.glob(f"{root_dir}/masks/*.png")) # Check extension!

# 2. Split Data
train_imgs, test_imgs, train_masks, test_masks = train_test_split(
    all_images, all_masks, test_size=0.2, random_state=42
)

# 3. Define Transform
transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.ToTensor()         # ToTensor does two things: Convert to tensor + scaling (divide by 255)
])
transform_mask = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.PILToTensor()       # PILToTensor does one thing: Convert to tensor only (mask should not be scaled!!)
])

# 4. Create Datasets
train_dataset = MulticlassDataset(train_imgs, train_masks, transform=transform, mask_transform=transform_mask)
test_dataset  = MulticlassDataset(test_imgs,  test_masks,  transform=transform, mask_transform=transform_mask)

# 5. Check Output
img, mask = train_dataset[0]
print(f"Img Shape: {img.shape}")   # [3, 256, 256]
print(f"Mask Shape: {mask.shape}") # [1, 256, 256]
print(f"Unique Classes: {torch.unique(mask)}")
```
